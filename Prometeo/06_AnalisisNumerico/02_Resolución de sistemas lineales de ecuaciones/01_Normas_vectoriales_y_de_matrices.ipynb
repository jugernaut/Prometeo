##Problema a resolver

Queremos encontrar la solución al problema de sistemas lineales:

$$ \begin{equation}
Ax = b
\end{equation}$$

para lo que necesitamos el concepto de norma:

**Norma vectorial**

Sea un entero p > 0 y un vector x, de dimension n. La norma p se
define como:

\begin{equation}
||x||_{p} = \left( \sum_{i= 1}^{n} |x_{i}|^{p} \right)^{1/p}
\end{equation}

La norma infinito

$$\lim_{p \to \infty} ||x||_{p} = ||x||_\infty = 1\leq i\leq n \{||x_i||\}$$

##Equivalencia de normas

Las normas son herramientas básicas para definir y analizar la
convergencia de una sucesión de vectores en espacios vectoriales. La
convergencia depende de las definición de norma que se usa. Podría
suceder que una sucesión de vectores convergiera para una norma, pero no
para otra. Pero en un espacio vectorial de dimensión finita todas las
normas son equivalentes en el siguiente sentido:

Dos normas \begin{equation} ||·|| , ||·||^{'} \end{equation} son equivalentes si existen m y M constantes reales positivas tales que para cualquier x tenemos
\begin{equation} m||x||  \leq  ||x||^{'} \leq M ||x|| \end{equation}

##Número de condición 
El número de condición de una matriz cuadrada no singular A con respecto a una norma matricial dada, se define como:
\begin{equation}cond(A) = ||A||·||A^{-1}||
\end{equation}

por convención cond(A) = $\infty$, si A es singular.

$\color{blue}{Ejemplo}$

La matriz A y su inversa estan dadas por:
\begin{equation} A=  \begin{bmatrix}
2 & -1 & 1\\
1 & 0 & 1\\
3 & -1 & 4
\end{bmatrix} , A^{-1}=  \begin{bmatrix}
0.5 & 1.5 & -0.5\\
-0.5 & 2.5 & -0.5\\
-0.5 & -0.5 & 0.5
\end{bmatrix} 
\end{equation}

Tenemos que $||A^{-1}||_1 = 4.5$ y $||A^{-1}||_\infty = 3.5$ por lo tanto:
$cond_1(A)= ||A||_1 ||A^{-1}||_1 = 6(4.5) = 27$ y
$cond_\infty(A)= ||A||_\infty ||A^{-1}||_\infty = 8(3.5) = 28$

Para este ejemplo vemos como el valor numérico del número de
condición de una matriz de n x n depende de la norma que se use. Por
la equivalencia de las normas vectoriales, estos valores pueden diferir a
lo más una constante fija que depende de n y por lo tanto son
igualmente útiles.

##Cálculo de $||A^{-1}||$

Algo que puede ser complicado es el cálculo de $||A^{-1}||$. De las
propiedades de las normas, sabemos que si z es una solución a Az = y
entonces:
\begin{equation} ||z||=||A^{-1}y|| \leq ||A^{-1}||||y||\end{equation}
de donde
\begin{equation} \frac{||z||}{||y||} \leq ||A^{-1}||\end{equation}

y esta cota se alcanza para algún vector elegido de forma óptima $y$ tal
que el cociente $\frac{||z||}{||y||}$ es tan grande como fuerá posible, entonces se tendría una estimación razonable para $||A^{-1}||$.

$\color{blue}{Ejemplo}$

Consideremos la matriz:

\begin{equation} A=  \begin{bmatrix}
0.913 & 0.659 \\
0.457 & 0.330 
\end{bmatrix} 
\end{equation}

Si elegimos $y = [0, 1.5]^T$ , entonces $z = [-7780, 10780]^T$ de donde

\begin{equation} ||A^{-1}|| \approx \frac{||z||_1}{||y||_1} \approx 1.238 \times 10^4\end{equation}

de donde

\begin{equation} cond(A) = ||A||||A^{-1}|| \approx 1.370 \times 1.238 \times 10^4 = 1.696 \times 10^4\end{equation}

El vector y fue elegido de forma que produce el máximo posible de $\frac{||z||}{||y||}$ y por lo tanto un valor correcto para $||A^{-1}||$. Encontrar tal óptimo y puede ser prohibitivo en términos computacionales.

##Cotas de error

El número de condición también nos da cotas cuantitativas de las cotas
de error cuando se calculan las soluciones a un sistema de ecuaciones.
Sea x la solución a un sistema de ecuaciones Ax = b.

Sistema de ecuaciones perturbado

Sea $\hat{x}$ la solución al sistema: $A\hat{x} = b + \bigtriangleup b$ con un vector de términos
independientes (del lado derecho) perturbado. Si definimos $\bigtriangleup x = \hat{x} - x$ entonces tenemos:

\begin{equation} A\hat{x} = A(x+\bigtriangleup x) = Ax+A\bigtriangleup x = b + \bigtriangleup b \end{equation}

como $Ax=b$ debemos tener $A\bigtriangleup x=\bigtriangleup b$, de donde $\bigtriangleup x = A^{-1}\bigtriangleup b$

Error relativo máximo

El objetivo es encontrar el error más grande posible al hacer la
perturbación. Para esto necesito calcular el error relativo máximo, es
decir la diferencia entre la solución original x y la solución con
perturbación x:

\begin{equation} \frac{||\hat{x}-x||}{||x||} = \frac{||\bigtriangleup x||}{||x||}  \end{equation}

Tomando normas, obtenemos las desigualdades:

$||b||=||Ax||\leq ||A||||x||$, o $||x|| \geq \frac{||b||}{||A||} \Rightarrow \frac{1}{||x||}\leq \frac{||A||}{||b||}$

y

\begin{equation} ||\bigtriangleup x|| = ||A^{-1} \bigtriangleup b|| \leq ||A^{-1}||||\bigtriangleup b||\end{equation}

Combinando estas igualdades obtenemos:

\begin{equation} \frac{||\bigtriangleup x||}{||x||} \leq ||A^{-1}||·||\bigtriangleup b||\frac{||A||}{||b||} \end{equation}

Por definición $||A||·||A^{-1}|| = cond(A)$ de donde obtenemos la cota:

\begin{equation} \frac{||\bigtriangleup x||}{||x||} \leq cond(A)\frac{||\bigtriangleup b||}{||b||} \end{equation}

##Número de condición de una matriz: factor de “amplifiación”

El número de condición de una matriz nos da un factor de
“amplificación” que acota el cambio relativo máximo en el vector de
términos independientes del lado derecho de la ecuación:

Cambio relativo o perturbación en la matriz A

Tenemos un resultado similar para cambios o perturbaciones similares,
pero ahora en la matriz A. Si $Ax = b y (A + E) \hat{x} = b$ entonces:

\begin{equation} \bigtriangleup x= x-\hat{x}= A^{-1} (A\hat{x}-b)= -A^{-1}E\hat{x} \end{equation}

De la ecuación anterior tomamos las normas y obtenemos la
desigualdad:

\begin{equation} ||\bigtriangleup x|| \leq ||A^{-1}||·||E||||\hat{x}||
\end{equation}

que usando la definción de número de condición obtenemos la cota:

\begin{equation} \frac{||\bigtriangleup x||}{||x||} \leq cond(A)\frac{||E||}{||A||} \end{equation}

Aveces se deonota a ||E|| como la matriz de perturbación: $||\bigtriangleup A||$.

##Definición del residual

Para verificar la solución a un sistema de ecuaciones sustituimos en la
ecuación y vemos que tan cercano es el ajuste entre el lado izquierdo y
derecho de la solución.

$\color{blue}{Residual}$

El residual de una solución aproximada \hat{x} a un sistema de ecuaciones
Ax = b es la diferencia:

\begin{equation} r= b-A\hat{x} \end{equation}

Si A es no singular entonces en teoría el error: $||\bigtriangleup x|| = ||\hat{x} - x|| = 0$ si y solo si $||r|| = 0$. En la práctica estas cantidades no son
necesariamente pequeñas. Notemos que si multimplicamos el sistema
de ecuaciones Ax = b por una constante arbiraria diferente de cero no
se ve afectado.

##Residual

Por lo tanto al residual hay que multiplicarlo por esta constante y este se
puede hacer arbitrariamente grande o pequeño dependiendo del
escalamiento del problema.

Relación entre el error y el residual

Notemos que

\begin{equation} ||\bigtriangleup x|| = ||\hat{x}-x|| = ||A^{-1}(A\hat{x}-b)||= ||-A^{-1}r|| \leq ||A^{-1}||·||r||
\end{equation}

dividiendo ambos lados por $||\hat{x}||$ y usando la definición de cond(A)
tenemos que:

\begin{equation} \frac{||\bigtriangleup x||}{||\hat{x}||} \leq \frac{||A||||A^{-1}||||r||}{||A||||\hat{x}||} = cond(A) \frac{||r||}{||A||·||\hat{x}||}
 \end{equation}

Entonces un residual pequeño implica un error relativo en la solución
solo cuando la matriz A este bien condicionada.

$\color{blue}{Ejemplo}$

Consideremos el siguiente sistema de ecuaciones:

\begin{equation} Ax = \begin{bmatrix}
0.913 & 0.659 \\
0.457 & 0.330 
\end{bmatrix} \begin{bmatrix}
x_1 \\
x_2 
\end{bmatrix} = \begin{bmatrix}
0.254 \\
0.127
\end{bmatrix} =b
\end{equation}

Consideremos las dos soluciones aproximadas:

\begin{equation} \hat{x_1} =  \begin{bmatrix}
-0.0827 \\
0.5
\end{bmatrix} \hspace{0.5cm}  \hat{x_2} = \begin{bmatrix}
0.999 \\
-1.001
\end{bmatrix} =b
\end{equation}

Las normas y sus respectivos residuales son:

\begin{equation} ||r_1||=2.1\times 10^{-4} \hspace{0.5cm} ||r_2||=2.4\times 10^{-2} \end{equation}

¿Cuál será la mejor solución? \hat{x_1} tiene el residual más pequeño, pero la
solución exacta es $x = [1,-1]^T$ . La razón es que A esta mal
condicionada.
