### Ejemplo 1

Un estudiante de lenguas participar en un concurso internacional cuyo principal reto es el conocimiento del vocabulario de un cierto idioma. Por ello, es importante que el estudiante utilice un mtodo de estudio adecuado para recordar el significado del mayor nmero de palabras posible.

La <a href=https://es.wikipedia.org/wiki/Curva_del_olvido>curva del olvido</a> puede ayudar al estudiante a generar un plan de estudio adecuado. La funcin que define esta curva es la siguiente:

$$
R(t) = e^{-t/S}
$$

donde $R$ es cuanto recordamos, $S$ es la intensidad del recuerdo y $t$ el tiempo. Podemos definir $S \in (0,1]$, donde $1$ es la mxima intensidad de recuerdo y un valor cercano a $0$ corresponde a algo que no nos interesa nada.

**Observacin**: $S$ no puede ser exactamente $0$ por que en ese caso la funcin $R(t)$ no est definida.

La siguiente grfica muestra cmo decrecen nuestros recuerdos con el paso del tiempo.

```python
t = np.linspace(0, 7, 100)    # Codominio de la funcin, de 0 a 7 con 100 puntos
R = lambda t, S: np.exp(-t/S) # Definicin de la funcin del olvido

# Definimos algunos parmetros para la grfica
par = [{'title':'Funcin del olvido: $R(t) = e^{-t/S}$', 
        'xlabel':'$t$ [das]',
        'ylabel':'$R$'}
      ]

# Inicializamos el plano Cartesiano
graf = vis.planoCartesiano(par=par)

# Realizamos tres grficas para S=0.1, S=0.5 y S=0.9] de la curva del olvido
for S in [0.1, 0.5, 0.9]:
    graf.plot(x= t, y= R(t, S), par={'label':'S = {}'.format(S)})

# Mostramos la leyenda de las grficas
graf.legend()
graf.show()
```

Ser posible determinar cada cuanto tiempo el estudiante debe repasar las palabras para que no las olvide y pueda ganar el concurso? De qu depende esto?

Tomemos por ejemplo el caso de $S=0.9$ (curva naranja). En qu parte de la grfica el descenso es ms rpido?, en otras palabras, en qu parte de la grfica el olvido se incrementa?,Para conocer ese descenso, debemos calcular la pendiente y eso lo podemos hacer con la siguiente frmula:

$$
m = \frac{R(t_2) - R(t_1)}{t_2 - t_1}
$$

Como es obvio, este valor de la pendiente cambiar cuando $t_1$ y $t_2$ cambien. 

Definamos entonces $h = t_2 - t_1$ y $t = t_1$. 

Con ello podemos escribir:

$$
m(t) = \frac{R(t + h) - R(t)}{h}
$$

En esta ltima frmula vemos que la pendiente depende de $t$.

Vamos a calcular $m$ para $h = 1$ y $t = [0,1,2,3,4,5,6]$:

```python
# Vamos a trabajar con el siguiente valor de S
S = 0.9
h = 1.0
td = np.arange(0,8,h) # Definicin de las t = 0,1,2,...,6
m = np.zeros(len(td)) # Arreglo para almacenar las pendientes

# Calculamos la pendiente usando la frmula antes descrita
for i,x in enumerate(td):
    m[i] = (R(x + h, S) - R(x, S)) / h

# Ponemos la informacin en un DataFrame y la mostramos
tabla = pd.DataFrame(np.array([td, m, ]).T, columns = ['t', 'm(t)'])
tabla
```

Observamos que la pendiente es negativa, lo cual indica un decrecimiento. Tambin, la magnitud de la pendiente (su valor absoluto) disminuye conforme $t$ se incrementa. Es posible visualizar la pendiente uniendo los puntos con los que se realiz el clculo de la misma, y tambin se puede graficar como funcin del tiempo, veamos ambas cosas en las siguientes grficas:

```python
# Definimos algunos parmetros para la grfica
par = [{'title':'Funcin del olvido: $R(t) = e^{-t/S}$ para $S = 0.9$', 
        'xlabel':'$t$ [das]',
        'ylabel':'$R$'},
       {'title':'Pendiente $m(t)$', 
        'xlabel':'$t$ [das]',
        'ylabel':'$m$'},
      ]

# Inicializamos el plano Cartesiano para ver dos grficas
graf = vis.planoCartesiano(1,2, par=par, par_fig={'figsize':(14,5)})

# Graficamos la curva del olvido junto con rectas que representan las pendientes
graf.plot(1, x = t, y = R(t, S), par={'color':'C2'})
graf.plot(1, x = td, y = R(td,S), par={'lw':1,'color':'black', 'marker':'o','zorder':5})

# Graficamos la pendiente
graf.plot(2, x = td, y = m, par={'ls':'--', 'lw':1, 'color':'black', 'marker':'o','zorder':5})

graf.show()
```

En la grfica de la izquierda, la lnea negra nos muestra grficamente el cambio en la pendiente de la recta que une los puntos de la grfica de $R(t)$. En la grfica de la derecha, observamos el cmbio de la pendiente $m$ en funcin de $t$.

Lo que estamos observando es la razn de cambio de $R(t)$ en intervalos de tiempo de longitud $h = 1$. Esto es lo que expresa la frmula de $m(t)$. 

Ser posible calcular **la razn de cambio instantneo**? Para ello, esta razn debera calcularse en un solo instante de tiempo, lo cual implica que $t_1 = t_2 \Longrightarrow h = 0$, y esto nos lleva a que la frmula de $m(t)$ no est bien definida (divisin por cero!). Pero, qu pasa si $h$ se hace muy pequea? es decir:

$$
\lim_{h \to 0}  \frac{R(t + h) - R(t)}{h}
$$

## Definicin de derivada

En todos los libros de clculo encontrars la siguiente definicin y notacin para la derivada de la funcin $f(x)$:

$$ 
\frac{d f}{dx} = f^\prime(x)=\lim_{h \to 0} \frac{f(x + h) - f(x)}{h} 
$$

La derivada existe siempre y cuando exista el lmite. Puedes imaginar cuando este lmite no existe? Observe que en la definicin anterior se est calculando la pendiente de la funcin $f(x)$ en $x$, cundo es que esta pendiente no se puede calcular?

### Ejemplo 2

Consideremos la funcin $f(x) = x^3$ y apliquemos la definicin de derivada:

$$
\frac{d f}{dx} = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} = \lim_{h \to 0} \frac{(x + h)^3 - x^3}{h}
$$

Si expandimos los trminos del numerador obtenemos:

$$
\frac{d f}{dx} = \lim_{h \to 0} \, (3x^2 + 3 x h + h ^2)
$$

Al calcular el lmite de la derecha obtenemos:

$$
\frac{d f}{dx} = 3x^2
$$

Hemos calculado la derivada analtica de $f(x) = x^3$. Veamos como se ven $f(x)$ y $f^\prime(x)$:

```python
# Definimos algunos parmetros para la grfica
par = [{'title':'$f(x)$ y $f^\prime(x)$', 
        'xlabel':'$x$',
        'ylabel':'$f, f^\prime$'}
      ]

x = np.linspace(0,10,100) # Codominio de las funciones

# Inicializamos el plano Cartesiano
graf = vis.planoCartesiano(par=par)

# Graficamos la funcin y su derivada
graf.plot(x = x, y = x**3, par={'label':'$f(x)=x^3$'})
graf.plot(x = x, y = 3*x**2, par={'label':'$f^\prime(x)=3x^2$'})

graf.legend()
graf.show()
```

## Reglas de derivacin

En general es fcil calcular la derivada de cualquier funcin y existen reglas para hacerlo ms fcil.
 

**Regla de potencias**

Para cualquier nmero real $n$ si $f(x)= x^n$, entonces
$$
f(x)^\prime = n * x^{n-1}
$$

**Regla de la funcin constante**

Si $f(x)=c$ es una funcin constante, entonces
$$ 
f^\prime(x)=0 
$$

**Regla de la multiplicacin por constante**

Si $c$ es cualquier constante y $f(x)$ es diferenciable, entonces $g(x)= c f(x)$ tambin es diferenciable y su derivada es:
$$ 
g^\prime(x) = c f^\prime(x) 
$$

**Regla de suma y diferencia**

Si $f(x)$ y $g(x)$ son diferenciables, entonces $f(x) + g(x)$ y $f(x) - g(x)$ tambin son diferenciables y sus derivadas son:
$$ 
[f(x)+ g(x)]^\prime = f^\prime(x) + g^\prime(x) 
$$

$$ 
[f(x)- g(x)]^\prime=f^\prime(x) - g^\prime(x) 
$$

**Regla del producto**

Si $f(x)$ y $g(x)$ son funciones diferenciables, entonces $f(x)g(x)$ es diferenciable y su derivada es:
$$ 
[f(x) g(x)]^\prime= f(x)g^\prime(x) + g(x)f^\prime(x) 
$$

**Regla del cociente**

Si $f$ y $g$ son funciones diferenciables y $g(x) \neq 0$, entonces $f(x)/g(x)$ es diferenciable y su derivada es: 
$$  
\left[\frac{f(x)}{g(x)} \right]^\prime = \frac{f(x)g^\prime(x)- f^\prime(x)g(x) }{g(x)^2}
$$

**Regla  de la cadena**

Si la funcin $f(u)$ es diferenciable, donde $u = g(x)$, y la funcin $g(x)$ es diferenciable, entonces la composicin $y=(f \circ g)(x)= f(g(x))$ es diferenciable:
$$ 
f(g(x))^\prime = f^\prime(g(x)) \cdot g^\prime(x)  
$$

**Regla de LHpital**

Esta regla es utilizada en caso de indeterminaciones donde $f(x)$ y $g(x)$ son dos funciones continuas definidas en el intervalo $[a,b]$, derivables en $(a,b)$ y sea $c$ perteneciente a $(a,b)$ tal que $f(c)=g(c)=0$ y $g^\prime(x) \neq 0$ si $x \neq c$. Si existe el lmite $L$ de $f^\prime/g^\prime$ en $c$, entonces existe el lmite de $f(x)/g(x)$ (en $c$) y es igual a $L$. Por lo tanto:
$$ 
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f^\prime(x)}{g^\prime(x)} = L 
$$

**Derivadas de funciones trigonomtricas**

$$
\begin{eqnarray}
\sin^\prime(x) & = & \cos(x) \\ 
\tan^\prime(x) & = & \sec^2(x) \\
\sec^\prime(x) & = & \sec(x)tan(x) \\
\cos^\prime(x) & = & -\sin(x) \\
\cot^\prime(x) & = & -\csc^2(x) \\
\csc^\prime(x) & = & -\csc(x)cot(x)
\end{eqnarray}
$$



***
**Teorema de Rolle :**
Sea $a < b$ y suponga que $f : [a, b]  {\mathbb{R}}$ es derivable en $(a, b)$ y continua
en $[a, b]$ y $f(a) = f(b)$. Entonces $ x_0  (a, b)$ tal que $f^\prime(x_0) = 0$
***
Lo anterior quiere decir que, dadas las condiciones del teorema, es posible encontrar un punto de la funcin $f(x)$ dentro del intervalo $(a,b)$ donde la derivada, esto es la lnea tangente, es cero, en otras palabras, en ese punto la lnea tangente es horizontal.

## Derivadas de orden superior

Es posible obtener la derivada de la derivada, es decir, si tenemos una funcin $f(x)$ cuya derivada es $f^\prime(x)$, entonces podemos calcular la derivada a esta ltima funcin, para obtener $f^{\prime\prime}(x)$, a esta ltima funcin, si es que existe, se le conoce como la segunda derivada de $f(x)$. Tambin se puede denotar a la segunda derivada com $f^{(2)}(x)$.

En general, si $f(x)$ es derivable $k$ veces, entonces es posible obtener la $k$-sima derivada de dicha funcin, que se escribe como:
$$ 
\frac{d^kf(x)}{dx^k} = f^{(k)}(x)
$$

### Ejemplo 4

Calcular la cuarta derivada de $f(x)= 3x^4 + 2x^2 - 20$.
  
Para calcular la derivada solicitada de $f(x)$ se deben realizar los clculos recursivos desde la primera derivada de la funcin:

$$
\begin{eqnarray}
f^\prime(x) & = & 12x^3 + 4x \\
f^{(2)}(x) & = & 36x^2 + 4 \\
f^{(3)}(x) & = & 72x  \\
f^{(4)}(x) & = & 72 \\
\end{eqnarray}  
$$ 

Observe las grficas de cada funcin y explique el comportamiento de cada una de ellas.

```python
# Definimos la funcin y sus cuatro derivadas
f = lambda x: 3*x**4 + 2*x**3 -20
f1 = lambda x: 12*x**3 + 4*x
f2 = lambda x: 36*x**2 + 4
f3 = lambda x: 72*x
f4 = lambda x: 72*np.ones(len(x))

x = np.linspace(-3, 3, num=50) # Codominio de la funcin

# Definimos algunos parmetros para la grfica
par = [{'title':'$f(x)=3x^4 + 2x^3 -20$ y sus derivadas', 
        'xlabel':'$x$',
        'ylabel':''}
      ]

# Inicializamos el plano Cartesiano
graf = vis.planoCartesiano(par=par)

# Graficamos la funcin y su derivada
graf.plot(x = x, y = f(x), par={'label':'$f(x)$'})
graf.plot(x = x, y = f1(x), par={'label':'$f^{(1)}(x)$'})
graf.plot(x = x, y = f2(x), par={'label':'$f^{(2)}(x)$'})
graf.plot(x = x, y = f3(x), par={'label':'$f^{(3)}(x)$'})
graf.plot(x = x, y = f4(x), par={'label':'$f^{(4)}(x)$'})
graf.legend()
graf.show()
```
