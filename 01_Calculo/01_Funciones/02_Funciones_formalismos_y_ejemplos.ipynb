Una definicin ms formal de funcin es como sigue:

Sean $A$ y $B$ dos conjuntos. Una funcin $f:A \rightarrow B$ es una relacin $f \subseteq A$ x $B$, es decir, $f$ define un conjunto de tuplas $(x,y)$, con las siguientes propiedades:

(i) Para toda $x \in A$, existe $y \in B$ tal que $(x,y) \in f$. <br>
**Nota**: En algunos textos encontrars una notacin un poco ms compleja como la siguiente:
$\forall x \in A, \exists y \in B \wedge (x,y) \in f$, la cual se lee as "*para toda $x$ que est en el conjunto $A$, existe una $y$ que est en el conjunto $B$ tal que la tupla $(x,y)$ est en el conjunto definido por $f$*".

(ii) Cada elemento de $A$ tiene asociado slo un elemento de $B$. <br>
**Nota**: Otra manera de expresar lo mismo es: $\forall x \in A$ $\forall y_1,y_2 \in B [((x,y_1) \in f \wedge (x,y_2) \in f) \Rightarrow y_1 = y_2 ]$



Juntando las condiciones (i) y (ii), podemos decir que para toda $x \in A$
existe una nica $y \in B$ tal que $(x,y) \in f$ . Entonces, si $f$ es una funcin de $A$ en $B$, para cada $x \in A$ denotamos con $f(x)$ al nico elemento de $B$ que le corresponde. Es decir, si $(x,y) \in f$, entonces escribimos $f(x) = y$. A $f(x)$ le llamamos la **imagen** ($im$) de $x$ bajo $f$. 

En ocasiones, decimos que el elemento $x \in A$ es mandado bajo la funcin $f$ a $f(x)$ y lo denotamos por $x \mapsto f(x)$. Como una funcin $f : A \rightarrow B$ es una relacin binaria entre $A$ y $B$, la definicin de la imagen de $f$ es:

$im(f) = \{b \in B : \exists a \in A \wedge (a,b) \in f\}$,
pero entonces, como $f$ es funcin,

$im(f)= \{y \in B : \exists x \in A f(x) = y \} = \{f(x) : x \in A\}$.
Por lo tanto, $im(f) \subseteq  B$.

**Observemos lo siguiente**: <br>
Si $y \in [2, \infty)$ entonces se cumple que $4 \le y^2$. Pero dado que $y = \sqrt{17x^2 + 4}$ entonces: $y^2 = 17x^2 + 4$, y despejando a $x$ obtenemos que $x = \pm \sqrt{\frac{y^2}{17} - \frac{4}{17}}$, y si a esta $x$ le aplicamos $f$ tenemos:

Por lo tanto $y \in im(f)$.

## Propiedades de las funciones

Sean $A, B$, $C$ y $D$ conjuntos cualesquiera. Sean
$f: A \rightarrow B$, $g:C \rightarrow D$. Definimos la siguientes operaciones entre $f$ y $g$:

1. La suma de $f$ y $g$ mediante la ecuacin: $(f + g)(x) = f(x) +g(x)$, donde
 $f+g:A\cap C \rightarrow B\cap D$ 
2. La resta de $f$ y $g$ mediante la ecuacin: $(f - g)(x) = f(x) -g(x)$, donde
 $f-g:A\cap C \rightarrow B\cap D$ 
3. El producto de $f$ y $g$ mediante la ecuacin: $(f \cdot g)(x) = f(x)\cdot g(x)$. Particularmente, si $f(x) = k$, definimos una nueva funcin $k\cdot g$ como $(k\cdot g)(x) = k\cdot g(x)$. Por lo tanto, $f \cdot g: A \cap C \rightarrow B\cap D$.
4. El cociente de $f$ y $g$ mediante la ecuacin: $\left(\frac{f}{g}\right)(x) = \frac{f(x)}{g(x)}$, siempre que $g(x) \neq 0$ para toda $x \in A$. Por lo tanto, $\displaystyle \frac{f}{g}: A \cap C \cap \{x|g(x)\neq 0\} \rightarrow B\cap D$.


**Teorema** Sean $A, B$, $C$,$D$, $E$ y $F$ conjuntos cualesquiera. Sean
$f:A\rightarrow B$, $g:C \rightarrow D$ y $h:E \rightarrow E$. Se cumple lo siguiente:

1. *Asociatividad*: $(f+g)+h=f+(g+h)$  y  $(f\cdot g)\cdot h=f\cdot (g\cdot h)$
2. *Conmutatividad*: $f+g=g+f$  y  $f\cdot g=g\cdot h$


## Tipos de Funciones

Una funcin $f : A \rightarrow B$ es **inyectiva** o **uno a uno** si y
slo si para cualesquiera $x_1, x_2 \in A$ con $x_1 \neq x_2$ se tiene que $f(x_1) \neq f(x_2)$, es decir:
$\forall x_1,x_2 \in A(x_1 \neq x_2 \Rightarrow f(x_1) \ne f(x_2))$.
Equivalentemente, usando la contrapuesta, $f$ es inyectiva si y slo si:

$\forall x_1,x_2 \in A(f(x_1) = f(x_2) \Rightarrow x_1 = x_2)$.

En otras palabras, una funcin es inyectiva si elementos distintos del dominio van a dar a elementos distintos del codominio al aplicar la funcin, o bien si la nica manera de que $x_1$ y $x_2$ en el dominio vayan a dar al mismo elemento del codominio sea que desde un principio $x_1 = x_2$.

Una funcin $f : A \rightarrow B$ es **sobreyectiva** o **suprayectiva** si y slo si para todo $y \in B$ existe $x \in A$ tal que $f(x) = y$ , es decir,
$\forall y (y \in B \Rightarrow \exists  x \in A \wedge f(x) = y$.
De acuerdo a lo anterior las funciones suprayectivas son aquellas en que todo elemento del codominio proviene de alguno del dominio.

Una funcin $f : A \rightarrow B$ es **biyectiva** si y slo si $f$ es
inyectiva y es sobreyectiva.

### Ejemplo 4

Sean $A = \{-1, 1, 2,3\}, B = \{-8, 4, 5, 8\}$ y $f:A\rightarrow B$
definida como $f(x) = -x^2 +2x +7$ .
Podemos ver que $f$ no es inyectiva, pues $f(-1) = 4$ y $f(3) = 4$, pero $-1 \neq 3$.
Adems, $f$ no es sobreyectiva, pues $5 \in B$ y no hay $a \in A$ tal que $f(a) = 5$. Sin embargo, si escogemos a los conjuntos $C$ y $D$ de tal forma que $C=A-\{3\}$ y $D=B-\{5\}$, la funcin $f:C\rightarrow D$ definida como $f(x) = -x^2 +2x +7$ s sera sobreyectiva e inyectiva, es decir, $f(x)$ sera biyectiva.

**NOTA:** Siempre es posible restringir el dominio de una funcin para crear una funcin inyectiva y el codominio para crear una funcin suprayectiva.

```python
# Dominio: creamos un arreglo de numpy con 4 elementos: -1,1,2,3
x1 = np.array([-1,1,2,3])
# Dominio: creamos un arreglo de numpy con 3 elementos: -1,1,2
x2 = np.array([-1,1,2])
# Codominio: a cada elemento del arreglo x1 le aplicamos la funcin -x^2 + 2x + 7
y1 = -x1**2+2*x1+7
# Codominio: a cada elemento del arreglo x2 le aplicamos la funcin -x^2 + 2x + 7
y2 = -x2**2+2*x2+7

# parmetros que contendr la prxima figura que vamos a crear y mostrar: 
# los dos ttulos, la etiqueta de los dos ejes x y la etiqueta de los dos ejes y
par = [{'title':'Ejemplo 1: Funcin que no es biyectiva', 
        'xlabel':'$X_1$',
        'ylabel':'$Y_1$'},
       {'title':'Ejemplo 1: Funcin convertida a biyectiva', 
        'xlabel':'$X_2$',
        'ylabel':'$Y_2$'}
      ]

# instanciamos una figura que contiene horizontalmente dos planos cartesianos
graf = vis.planoCartesiano(1,2,par=par)
# graficamos en el primer plano de la figura un diagrama de dispersin para mostrar 
# la funcin que no es biyectiva
graf.scatter(1, x = x1, y = y1, par={'color':'purple', 's':50})
# graficamos en el segundo plano de la figura un diagrama de dispersin para mostrar 
# la funcin que convertida a biyectiva
graf.scatter(2, x = x2, y = y2, par={'color':'blue', 's':50})
# mostramos la figura que contiene la grfica de cada plano 
graf.show()
```

### Ejemplo 5

Sea $f : \mathbb{R}^+ \rightarrow (-8,\infty)$ dada por $f(x) = 9x -8$. Vamos a mostrar que $f$ es biyectiva:

1. Sean $x, y  \in \mathbb{R}^+$ tales que $f(x) = f(y)$. Entonces: $9x -8 = 9y -8 \Longleftrightarrow 9x = 9y \Longleftrightarrow x = y$. Por lo tanto, $f$ es inyectiva. 

2. Ahora bien, sea $y \in (-8,\infty)$. Necesitamos encontrar $x \in \mathbb{R}^+$ tal que $f(x)=y$. Dado que $f(x)=9x -8$, necesitamos que $9x -8= y$, entonces $\displaystyle x =\frac{y + 8}{9}$. Entonces $\displaystyle f(x) = f (\frac{y + 8}{9}) = 9(\frac{y + 8}{9}) -8 = (y+8)-8= y$. Por lo tanto, para cualquier $y \in (-8,\infty)$, hay una $x \in \mathbb{R}^+$  tal que $f (x) = y$, entonces $f$ es sobreyectiva.

Por lo tanto, los puntos 1. y 2. indican que $f$ es biyectiva. Veamos la grfica:

```python
# creamos un arreglo de numpy con 2 elementos: 0 y 5. Para esta grfica, el dominio
# va ir de 0 a 5
x3 = np.array([0,5])
# Codominio: a los elementos del arreglo x3 les aplicamos la funcin 9x-8
y3 = 9*x3 - 8

# parmetros que contendr la prxima figura que vamos a crear y mostrar: 
# el ttulo, la etiqueta del eje x y la etiqueta del eje y
par = [{'title':'Grfica de la recta $9x-8$', 
        'xlabel':'$X_3$',
        'ylabel':'$Y_3$'}
      ]

# instanciamos una figura que bsicamente es el plano cartesiano
graf = vis.planoCartesiano(par=par)
# graficamos en la figura la funcin f(x)=9x-8 (que es una recta) con el dominio que creamos 
graf.plot(1, x = x3, y = y3, par={'label':'$f(x)=9x-8$'})
# mostramos la figura que contiene la grfica de la recta
graf.show()
```

**Observacin:** cuando la grfica de una funcin se puede representar mediante un sistema de coordenadas cartesianas en el plano, a partir de la grfica podemos verificar si la funcin es inyectiva, ya que si no lo es existen $x_1 \ne x_2$ en el dominio de la funcin tales que $f(x_1) = f(x_2)$ y en la grfica apareceran los puntos $(x_1 , y)$ y
$(x_2 , y)$ donde $y = f (x_1) = f (x_2)$, que tienen distinta abscisa pero la misma ordenada. Podemos entonces trazar rectas horizontales y si existe alguna que corte a la grfica en ms de un punto, la funcin no es inyectiva.

## Composicin de Funciones

Sean $A, B$ y $C$ conjuntos cualesquiera. Sean $B' \subseteq B$,
$f:A\rightarrow B'$ y $g:B \rightarrow C$. Definimos la **composicin** de $f$ y $g$, denotada por $g \circ f$ , como la funcin $g \circ f : A \rightarrow C$ tal que para toda $x \in A$  se tiene $(g \circ f )(x) =
g(f(x))$.

### Ejemplo 7
Sean $f : \mathbb{R} \rightarrow  \mathbb{R}$ y $g : \mathbb{R} \rightarrow \mathbb{R} $ definidas por $f (x) = 5x^2 + 4$
y $g(x) = 8x -3$.
Entonces la composicin $g \circ f: \mathbb{R} \rightarrow \mathbb{R}$ est dada por $(g \circ f )(x) =
g(f (x)) = g(5x^2 + 4) = 8(5x^2 + 4) -3 = 40x^2 + 32 -3 = 40x^2 + 29 $.

La composicin $f \circ g: \mathbb{R} \rightarrow \mathbb{R}$ est dada por $(f \circ g)(x) = f (g(x)) =f(8x -3) = 5(8x -3)^2 + 4 = 5(64x^2 - 48x +9)+4 $ $= 320x^2 -240x + 45 + 4 = 320x^2 -240x + 49$.

Observamos que las funciones $g \circ f$ y $f \circ g$ son diferentes, por lo tanto en general, la composicin de funciones no es conmutativa.

Las siguientes grficas muestran lo anterior:

```python
# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la funcin f(x)=5x^2+4. Si x es una lista, se vala a todos los elementos de dicha lista
def f5(x):
    return 5 * x**2 + 4

# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la funcin g(x) 8x-3. Si x es una lista, se vala a todos los elementos de dicha lista
def g5(x):
    return 8*x - 3 

# Dominio: creamos un rango de valores dentro del intervalo semiabierto [-10, 10), 
# que va del -10 al 10 con pasos de 0.1
x5 = np.arange(-10, 10, 0.1) 

# Realizamos la composicin de funciones f(g(x))
f_g = f5( g5(x5) ) 

# Realizamos la composicin de funciones g(f(x))
g_f = g5( f5(x5) )

# parmetros que contendr la prxima figura que vamos a crear y mostrar: 
# el ttulo y la etiqueta del eje x 
par = [{'title':'Composicin de funciones', 
        'xlabel':'$X_5$'},       
      ]
# instanciamos una figura que bsicamente es el plano cartesiano
graf = vis.planoCartesiano(par=par)
# graficamos en la figura la funcin f(g(x)) en el intervalo semiabierto que creamos: [-10, 10) con pasos de 0.1
# NOTA: Por defecto, el primer color de la establece lo coloca como azul
graf.plot(x = x5, y = f_g, par={'label':'$(f \circ g)(x)$'})
# graficamos en la figura la funcin g(f(x)) en el intervalo semiabierto que creamos: [-10, 10) con pasos de 0.1
# NOTA: Por defecto, el segundo color de la grfica lo establece como naranja
graf.plot(x = x5, y = g_f, par={'label':'$(g \circ f)(x)$'})
# indicamos que queremos mostrar en la figura la leyenda de las dos grficas
graf.legend()
# mostramos la figura que contiene las grficas que acabamos de generar
graf.show()
```

**Teorema.** Sean $A, B, C$ y $D$ conjuntos cualesquiera, y sean $B' \subseteq B$
y $C' \subseteq C$. Sean $f : A \rightarrow B'$ , $g : B \rightarrow C'$ y $h : C \rightarrow D$ funciones. Entonces
$h \circ (g \circ f ) = (h \circ g) \circ f$ , es decir, la composicin de funciones es asociativa.

### Ejemplo 8

Sean $f : \mathbb{R} \rightarrow \mathbb{R}$ , $g : \mathbb{R}\rightarrow \mathbb{R}$ y $h : \mathbb{R} \rightarrow \mathbb{R}$ funciones dadas por las siguientes reglas de correspondencia: $f(x) = e^x$, $g(x) = 3x^2$ y $h(x) = sin(x)$.

Por un lado, tenemos que: 

$h \circ (g \circ f )(x) = h \circ (g(f(x)) = h \circ g(e^x) = h( g(e^x)) = h( g(e^x)) = h(3e^{2x}) = sin(3e^{2x})$

Por otro lado:

$(h \circ g) \circ f = h(3x^2) \circ f = sin(3x^2) \circ f = sin(3(e^x)^2) = sin(3e^{2x})$

Por lo tanto $h \circ (g \circ f ) = (h \circ g) \circ f$ 

Veamos las grficas:

```python
# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la funcin h(x)=sin(x). Si x es una lista, se vala a todos los elementos de dicha lista
def h6(x):
    return np.sin(x)

# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la funcin g(x)=3x^2(x). Si x es una lista, se vala a todos los elementos de dicha lista
def g6(x):
    return 3*x**2

# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la funcin f(x)=e^x. Si x es una lista, se vala a todos los elementos de dicha lista
def f6(x):
    return np.exp(x)

# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la composicin de funciones h(g(x)). Si x es una lista, se vala a todos los elementos de dicha lista
def h_g(x):
    return h6(g6(x))

# creamos una funcin en python, la cual recibe como parmetro un elemento x. Retornamos ese elemento
# valuado en la composicin de funciones g(f(x)). Si x es una lista, se vala a todos los elementos de dicha lista
def g_f(x):
    return g6(f6(x))

# Dominio: creamos un rango de valores dentro del intervalo semiabierto [-3, 3.1), 
# que va del -3 al 3.1 con pasos de 0.1
x6 = np.arange(-3, 3.1, 0.1)

# Realizamos la composicin de funciones h  (g  f)
h_g_f_1 = h6( g_f(x6) )

# Realizamos la composicin de funciones (h  g)  f
h_g_f_2 = h_g( f6(x6))

# parmetros que contendr la prxima figura que vamos a crear y mostrar: 
# los dos ttulos, la etiqueta de los dos ejes x y la etiqueta de los dos ejes y
par = [{'title':'$h \circ (g \circ f)$', 
        'xlabel':'$x$',
        'ylabel':'$y$'},
       {'title':'$(h \circ g ) \circ f$', 
        'xlabel':'$x$',
        'ylabel':'$y$'}, 
      ]

# instanciamos una figura que contiene horizontalmente dos planos cartesianos
graf = vis.planoCartesiano(1, 2, par=par)
# graficamos en el primer plano de la figura la composicin de funciones h  (g  f)
graf.plot(1, x = x6, y = h_g_f_1, par={'color':'C0'})
# graficamos en el segundo plano de la figura la composicin de funciones (h  g)  f
graf.plot(2, x = x6, y = h_g_f_2, par={'color':'C1'})
# mostramos la figura que contiene la grfica de cada plano 
graf.show()
```

**Teorema**

$(i)$ La composicin de funciones inyectivas es inyectiva.

$(ii)$ La composicin de funciones sobreyectivas es sobreyectiva (siempre que el codominio de la primera sea igual al dominio de la segunda).

$(iii)$ La composicin de funciones biyectivas es biyectiva.

## Funciones inversas

- Definimos la **funcin identidad** como $id(x) = x$, es decir, la imagen de $x$ bajo $id$ es $x$. 

- Sea $f : A \rightarrow B$ una funcin real biyectiva. La **funcin inversa** (o funcin recproca) de $f$, denotada por $f^{-1}$, es aquella que hace el camino inverso, asignando a los elementos de $B$ elementos de $A$, definida como:

$f(x) = y \Longleftrightarrow f^{-1}(y) = x$ 

- La funcin $f^{-1}$, al igual que $f$, es tambin biyectiva y queda determinada de modo nico por $f$ y cumple:

    + $f^{-1} \circ f=id_{A}$ en este caso $f^{-1}$ es inversa por la izquierda.
    + $f \circ f^{-1}=id_{B}$ en este caso $f^{-1}$ es inversa por la derecha.

El siguiente teorema da la condicin necesaria y suficiente para que $f$
tenga inversa.

**Teorema** Sean $A$ y $B$ conjuntos cualesquiera y sea $f : A \rightarrow B$. Se cumple lo siguiente:

$(i)$ Si $A \neq \emptyset$, entonces $f$ es inyectiva si y slo si $f$ tiene inverso izquierdo.

$(ii)$ $f$ es sobreyectiva si y slo si $f$ tiene inverso derecho.

**Observaciones**

1. Las grficas de $ f(x) $ y $f^{-1}(x)$  son simtricas respecto de la bisectriz del primer y tercer cuadrante.
2. Para construir o calcular la funcin inversa de una funcin cualquiera, se deben seguir los siguientes pasos:

  2.1. Se escribe la funcin con $x$ e $y$.

  2.2. Se despeja la variable $x$ en funcin de la variable $y$.

  2.3. Se intercambian las variables.

```python
# creamos un diccionario, el cual tiene como llaves algunas funciones: sin(x), 
# cos(x), tan(x), floor(x), ceil(x), exp(x), sqrt(x) y constantes conocidas como pi y e; 
# sus correspondientes valores (que vamos a utilizar para reemplazarlas) son esas
# funciones desde las bibliotecas numpy y math
funciones = {'sin' : 'np.sin', 'cos' : 'np.cos', 'tan' : 'np.tan', 'floor': 'np.floor',
             'ceil': 'np.ceil','exp': 'np.exp','sqrt': 'np.sqrt', 'pi': 'math.pi',
             'e': 'math.e','^': '**',
}

# lista que contiene las palabras permitidas que puede introducir el usuario
palabras_permitidas = ['x','sin','cos','tan','ceil','floor','sqrt','exp',
                       'abs','pi','e']

# creamos la funcin en python stringAFuncion. Recibe como parmetro un string. Nos va a 
# permitir graficar una funcin introducida por el usuario, siempre y cuando est 
# permitido ese string
def stringAFuncion(string):
    ''' evala una cadena y devuelve una funcin de x '''
    # encuentra todas las palabras y verificamose si todas estn permitidas:
    # utilizamos una expresin regular 
    for word in re.findall('[a-zA-Z_]+', string):
      # si alguna palabra introducida por el usuario no est en lista palabras_permitidas,
      # le indicamos que est prohibida esa expresin matemtica
        if word not in palabras_permitidas:
            raise ValueError(
                '"{}" est prohibido usar en expresiones matemticas'.format(word)
            )

    # reemplazamos cada llave del diccionario nombrado como funciones para reemplazar su llave
    # por su valor
    for old, new in funciones.items():
        string = string.replace(old, new)
    # creamos la funcin de python func para valuar la funcin introducida como un string 
    def func(x):
        return eval(string)

    return func
```
