## Circunferencia

El concepto analítico de circunferencia puede ejemplificarse como un conjunto de puntos equidistantes a otro punto fijo llamado centro.

Aquella distancia entre los puntos y el centro se llama **radio**. Si consideramos el centro $C(h,k)$ y el punto $P(x,y)$ tenemos que la distancia entre $C$ y $P$ es el radio de la circunferencia y esta dado por:
$$ r= \sqrt ( (x-h)^2+(y-k)^2) $$

Derivada de la expresión anterior, se obtiene la forma ordinaria de la circunferencia:
$$ r^2= (x-h)^2+(y-k)^2 $$



```python
#Puntos a graficar
theta = np.linspace(0, 2*np.pi, 100)
#Radio
radius = 3
#Ecuacion de circulo
a = radius*np.cos(theta)
b = radius*np.sin(theta)

figure, axes = plt.subplots(1,figsize=(10,6))

#Ejes y limites
axes.set_xlim(-5, 5)
axes.set_ylim(-5,5)
axes.set_aspect(1)
axes.plot([-5,5],[0,0],color='black', linewidth=1)
axes.plot([0,0],[-5,5],color='black',linewidth=1)

#Circunferencia
axes.plot(a, b)

#Radio
axes.arrow(0,0,2.5,0,width=0.1, ec='red')
plt.text(0.7,-0.5, 'radio',size=15)
plt.title('Circunferencia')
plt.show()
```

## Parábola
Una parábola es el lugar geométrico de todos los puntos del plano cuya distancia no dirigida a un punto fijo llamado **foco** es igual a la distancia no dirigida a una recta fija llamada **directriz**. 

Partiendo de esta definición, podemos obtener la ecuación que representa a una parábola, para ello tendremos en cuenta dos tipos de parábolas: las verticales y las horizontales, empecemos con las parábolas verticales.

Se aplican por ejemplo, en algunos tipos de lentes y espejos, en puentes, trayectorias de proyectiles y balones.







```python
#Parámetros para la parabola
a = 1
b = 0
c = 0

#Definimos la función f1(x)
f1 = lambda x: ((a * x) ** 2) + (b * x) + c
f2 = lambda x: -15+x*b

#Definimos el dominio de la función f(x)
x = np.linspace(-10, 10, 100)

#Graficamos especificando el título y nombre de los ejes
par = [{'title':r"Parábola", 
        'xlabel':'$x$',
        'ylabel':'$y$'}
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=f1(x), par={'label':'$f1(x) = ax^2+bx+c$'})               #Gráfica de la función f1(x)
graf.plot(x=[0,0], y=[0,15], par={'label':'$p$', 'ls':'--','zorder':1 })
graf.plot(x=x, y=f2(x), par={'label':'$Directriz$', 'ls':'--','zorder':1 })

graf.scatter(x=0, y=0,        #vértice
             par={'label':'Vértice de la parábola', 's':100, 'color':'C1', 'zorder':10})
graf.scatter(x=0, y=15,        #foco
             par={'label':'Foco de la parábola', 's':100, 'color':'C2', 'zorder':10})

graf.legend()
graf.show()
```

Ejercicio: Cambiar los parámetros $a,b,c$ de la gráfica anterior para ver el comportamiento de la parábola.

Los elementos de la parábola son los siguientes:

*   $V(h,k):$ Vértice de la parábola en las coordenadas $(h,k)$.
*   $F(h,k+p):$ Foco de la parábola
*   $P(x,y):$ Punto en la parábola en las coordenadas (x,y).
*   $p:$ Distancia entre el vértice y el foco.
*   $y=k-p$: Ecuación de la directriz.

La ecuación ordinaria para la parábola es: 

$$ (x-h)^2=\pm4p(y-k)$$

Dada esta ecuación notemos que si:
*   $4p>0$ la parábola es cóncava hacia arriba.
*   $4p<0$ la parábola es cóncava hacia abajo.


Por otro lado la ecuación de una parábola horizontal es:
$$ (y-k)^2=\pm4p(x-h)$$

Dada esta ecuación notemos que si:
*   $4p>0$ la parábola es cóncava hacia la derecha.
*   $4p<0$ la parábola es cóncava hacia la izquierda.

```python
#Parámetros para la parabola
a = -1
a2 = 1
b = 0
c = 0

#Definimos la función f1(x)
f1 = lambda x: (a * x** 2) + (b * x) + c
f2 = lambda x: (a2 * x** 2) + (b * x) + c

#Definimos el dominio de la función f(x)
x = np.linspace(-10, 10, 100)

#Graficamos especificando el título y nombre de los ejes
par = [{'title':r"Parábolas", 
        'xlabel':'$x$',
        'ylabel':'$y$'}
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=f1(x), par={'label':'$f1(x) = ax^2+bx+c$'})              #Gráfica de la función f1(x)
graf.plot(x=x, y=f2(x), par={'label':'$f1(x) = -ax^2+bx+c$'})             #Gráfica de la función f2(x)


graf.legend()
graf.show()
```

## Elipse

Se define como el lugar geométrico de los puntos del plano en los que la suma de las distancias dirigidas a dos puntos fijos es constante.


*   El punto medio del segmento que une los focos es el **centro** de la elipse.
*   El segmento que une los puntos de intersección de la elipse con la recta que une los focos es el **eje mayor** y esos puntos de intersección son los vértices. 
*   El segmento perpendicular al eje mayor que pasa por el centro y acotado por la elipse es el **eje menor**. 
*   El segmento perpendicular al eje mayor que pasa por los focos, acotado por la elipse, es el **lado recto**.


```python
#Parámetros
u = 1.0    #coordenada x del centro
v = 0.5    #coordenada y del centro
a = 2.0    #radio en eje x
b = 1.5    #radio en eje y

#Definimos las funciones
f1 = lambda x: u+a*np.cos(x)
f2 = lambda x: v+b*np.sin(x)

par = [{'title':r"Elipse", 
        'xlabel':'$x$',
        'ylabel':'$y$'}
      ]

graf = vis.planoCartesiano(par=par)

#Puntos en x
x = np.linspace(0, 2*np.pi, 100)
graf.plot(x=f1(x), y=f2(x))

#Eje mayor
graf.plot(x=[-1,3], y=[0.5,0.5], par={'label':'Eje mayor', 'ls':'--', 'color':'C2', 'zorder':10})
#Eje menor
graf.plot(x=[1,1], y=[-1,2], par={'label':'Eje menor', 'ls':'--', 'color':'C3', 'zorder':10})
#Lados rectos
graf.plot(x=[2.5,2.5], y=[-0.5,1.5], par={'label':'Lado recto', 'ls':'--', 'color':'C4', 'zorder':10})
graf.plot(x=[-0.5,-0.5], y=[-0.5,1.5], par={'label':'', 'ls':'--', 'color':'C4', 'zorder':10})

graf.scatter(x=u, y=v,        #Centro
             par={'label':'Centro de elipse', 's':200, 'color':'C1', 'zorder':10})

graf.scatter(x=4, y=0,   
             par={'s':5, 'color':'C1', 'zorder':10})
             
graf.legend()
graf.show()
```

Si se tienen elipses **verticales** la ecuación será de la forma:
$$\frac{x^2}{a^2}+\frac{y^2}{b^2}=1$$

Si el centro esta fuera del origen:
$$\frac{(x-h)^2}{a^2}+\frac{(y-k)^2}{b^2}=1$$

Si se tienen elipses **horizontales** la ecuación será de la forma:
$$\frac{x^2}{b^2}+\frac{y^2}{a^2}=1$$

Si el centro esta fuera del origen:
$$\frac{(x-h)^2}{b^2}+\frac{(y-k)^2}{a^2}=1$$

## Hipérbola
Se define como el lugar geométrico de los puntos del plano en los que la diferencia de las distancias dirigidas a dos puntos fijos es constante.

La ecuación ordinaria de la hiperbola es:
$$\frac{x^2}{b^2}-\frac{y^2}{a^2}=1$$

Si el centro esta fuera del origen:
$$\frac{(x-h)^2}{b^2}-\frac{(y-k)^2}{a^2}=1$$

Para el lado recto de las hiperbolas, la expresión es:

$$LR = \frac{2b^2}{a}$$

Donde 

*   $a:$ Representa la distancia del centro de la hipérbola a uno de los vértices.
*   $b:$ Representa la distancia del centro a un extremos cualquiera del eje conjugado.
*   $c:$ Representa la distancia del centro a uno de los focos.

##Ecuación general de la hipérbola

La ecuación esta dada por: 

$$  A x^2+ Cy^2 + Dx + Ey + F = 0$$

Donde $A$ y $C$ tienen signos diferentes y son también diferentes de cero.

Si la ecuación de una hipérbola está expresada de manera general, se puede obtener su fórmula reducida utilizando el método de completar dos trinomios cuadrados perfectos.



```python
#Parámetros
a = 1    #radio en eje x
b = 1.5    #radio en eje y

#Definimos las funciones
f1 = lambda x: (np.sqrt(x**2*a**2 - a**2 ))
f2 = lambda x: -np.sqrt(x**2*a**2 - a**2 )

par = [{'title':r"Hipérbola", 
        'xlabel':'$x$',
        'ylabel':'$y$'}
      ]

graf = vis.planoCartesiano(par=par)

#Puntos en x
x = np.linspace(-10,10, 50)
graf.plot(x=x, y=f1(x),par={'color':'C0', 'zorder':10})
graf.plot(x=x, y=f2(x),par={'color':'C0', 'zorder':10})
#asintotas
graf.plot(x=x, y=x,par={'label':'Asíntotas','ls':'--','color':'C1', 'zorder':10})
graf.plot(x=x, y=-x,par={'ls':'--','color':'C1', 'zorder':10})
#Focos
graf.scatter(x=1, y=0,   
             par={'label':'Focos','s':100, 'color':'C2', 'zorder':10})
graf.scatter(x=-1, y=0,   
             par={'s':100, 'color':'C2', 'zorder':10})
graf.legend()
graf.show()
```

##Ecuaciones generales de segundo grado
Su forma es la siguiente
$$  A x^2+ Bxy + Cy^2 + Dx + Ey + F = 0$$

Es posible probar que toda ecuación de segundo grado corresponde a una cónica. El valor $B^2 + 4AC$ permite identificar la cónica representada por la ecuación.

Los criterios son los siguientes:

*   Si  $B^2 + 4AC<0$ la ecuación representa una elipse.
*   Si  $B^2 + 4AC=0$ la ecuación representa una parábola.
*   Si  $B^2 + 4AC>0$ la ecuación representa una hipérbola.

Al número $B^2 + 4AC$ se le llama **discriminante**.

## Traslación de ejes coordenados
La ecuación de toda curva depende del sistema de ejes coordenados que se tomen como referencia. Por lo que una misma curva puede tener ecuaciones cartesianas distintas dependiendo de los ejes coordenados seleccionados. 

En muchas ocasiones es posible simplificar las ecuaciones de las curvas, realizando traslaciones de los ejes coordenados.

Se debe realizar la siguiente transformación para trasladar una curva del eye $xy$ al eje $x'y'$.

Reemplazaren la ecuación: 
$x$ por $x'+h$ y $y$ por $y'+k$

Esto se aplica cuando el sistema de ejes $x'y'$ es paralelo al sistema de ejes $xy$ y el nuevo origen esta dado por el punto $(h,k)$.


###Ejemplo 2
Obtener la ecuación de la curva $ x^2+y^2-6x-4y+12=0$ trasladada al origen en el punto $O(3,2)$.

Es necesario sustituir los valores de $x$ por $x'+3$ y los valores de $y$ por $y'+2$ 

\begin{align}
x^2+y^2-6x-4y+12&=0\\
(x'+3)^2+(y'+2)^2-6(x'+3)-4(y'+2)+12&=0\\
x'^2 +6x'+9+y'^2+4y'+4-6x'-18-4y'-8+12&=0\\
x'^2+9+y'^2+4-18-8+12&=0\\
x'^2+y'^2&=1\\
\end{align}

La ecuación de la recta trasladada al sistema de ejes $x'y'$ es $x'^2+y'^2=1$

##Rotación de ejes

Otra transformación que se puede realizar es la rotación de ejes en la cual el origen de los sistemas de ejes de $xy$ y $x'y'$ es el mismo y se tiene un parámetro $\alpha$ que indica el ángulo en el que girarán los ejes $xy$. 

La rotación es la siguiente:

Se debe reemplazar $x$ por $x' cos(\alpha) -y' sen(\alpha)$ 

Y reemplazar  $y$ por $y' cos(\alpha) +x' sen(\alpha)$

Cuando realizamos traslación es posible eliminar terminos lineales, a diferencia de la rotación que nos permite simplificar la ecuación de la curva eliminando los términos mixtos $Bxy$.

Los criterios para elegir el ángulo de rotación y eliminar términos mixtos es:

*   Si $A=C$  entonces $\alpha=45°$
*   Si $A\neq C$  entonces $tan (2\alpha) = \frac{B}{A-C}$

### Ejemplo 3

Obtener la ecuación de la curva $ 41x^2-18xy+41y^2=800$ después de realizar una rotación de $45°$.

Sustituimos los nuevos valores para $x$ y para $y$
$$x=x' cos(45°) -y' sen(45°) = \frac{x'}{\sqrt2}-\frac{y'}{\sqrt2}$$

$$y= y' cos(45°) +x' sen(45°) = \frac{y'}{\sqrt2}+\frac{x'}{\sqrt2}$$

Ahora sustituyendo en la ecuación de la curva:
\begin{align}
41x^2-18xy+41y^2&=800\\
41(\frac{x'}{\sqrt2}-\frac{y'}{\sqrt2})^2-18(\frac{x'}{\sqrt2}-\frac{y'}{\sqrt2})(\frac{y'}{\sqrt2}+\frac{x'}{\sqrt2})+41(\frac{y'}{\sqrt2}+\frac{x'}{\sqrt2})^2&=800\\
41x^2-18xy+41y^2&=800
\end{align}

Simplificando obenemos que la ecuación queda
$$ 16x'^2 +25y'^2 = 400$$
