La trigonometría es la rama de las matemáticas que estudia la relación entre los lados y ángulos de los triángulos. Se ocupa, de las funciones asociadas a los ángulos, denominadas funciones trigonométricas.

Etimológicamente, trigonometría significa medida de los triángulos, ya que proviene de las palabras griegas trigono (triángulo) y metría (medida).

La trigonometría tiene innumerables aplicaciones en diversos campos de la ciencia: de una u otra manera en todos los campos de las matemáticas; en la física, por ejemplo, en fenómenos ondulatorios, medir distancias entre dos ubicaciones; en la astronomía, por ejemplo, para medir distancias entre planetas, navegación satelital, etc.

## Tipos de ángulos

Existen distintos tipos de angulos

* Un ángulo que mide entre 0° y 90° se llama un **ángulo agudo**.

*  Un ángulo que mide exactamente 90° es un **ángulo recto**.

*  Un ángulo que mide más de 90° pero menos de 180° es un **ángulo obtuso**.

* Un ángulo de exactamente 180° es un **ángulo llano**.

## Radianes
Existen tres unidades que emplea la trigonometría para la medición de ángulos: el radián, el gradián o grado centesimal (que permite dividir la circunferencia en cuatrocientos grados centesimales) y el grado sexagesimal (se usa para dividir la circunferencia en trescientos sesenta grados sexagesimales).

En la mayoría de las aplicaciones de trigonometría, los ángulos se miden en grados. En cursos más avanzados de matemáticas se prefiere que los ángulos se midan en radianes. Esta medida permite tratar a las funciones trigonométricas como funciones con dominios de números reales, en lugar de ángulos.


###Definición de radián
Un ángulo cuyo vértice se ubique en el centro de un círculo que intercepta un arco en el círculo de igual longitud al radio de éste tiene una medida de 1 **radián**.


```python
#Puntos a graficar
theta = np.linspace(0, 2*np.pi, 100)
#Radio
radius = 3
#Ecuacion de circulo
a = radius*np.cos(theta)
b = radius*np.sin(theta)

figure, axes = plt.subplots(1,figsize=(10,6))

axes.set_xlim(-5, 5)
axes.set_ylim(-5,5)
axes.set_aspect(1)
axes.plot([-5,5],[0,0],color='black', linewidth=1)
axes.plot([0,0],[-5,5],color='black',linewidth=1)
axes.plot(a, b)

axes.arrow(0,0,4,0,width=0.1, ec='red')
axes.arrow(0,0,3,2.5,width=0.1, ec='red')
plt.text(3,1, 'r',size=15)
plt.text(1.9,-0.5, 'r',size=15)
plt.annotate(r'$\theta$ = 1 radián',
            xy=(1.5, 0.5), xycoords='data',
            xytext=(-10,100), textcoords='offset points',
            size=15,
            arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"))

plt.title('Radián')
plt.show()
```

### Conversión entre grados y radianes

Recordemos que la circunferencia de un círculo esta dada por $2 \pi r$. Por lo tanto, un ángulo de 360°, que corresponde a un círculo completo, intercepta un arco de longitud igual a 
$2\pi$ veces el radio del círculo. Por consiguiente, un ángulo de 360° tiene una medida de $2\pi$  radianes:
$$ 360º = 2\pi* radianes$$

Por lo tanto obtenemos que:
$$ 1º = \frac{\pi}{180} radián$$
y
$$ 1 radián = \frac{180º}{\pi}$$


Las razones trigonométricas de un ángulo α son las razones obtenidas entre los tres lados de un triángulo rectángulo. Es decir, la comparación por su cociente de sus tres lados a, b y c.



*   El seno de un ángulo α se define como la razón entre el cateto opuesto  *lado a* y la hipotenusa *lado c*.

$$sen( \alpha) = \frac{cateto Opuesto}{hipotenusa}=\frac{a}{c}$$

*   El coseno se define como la razón entre el cateto contiguo o cateto adyacente  *lado b* y la hipotenusa  *lado c*

$$cos( \alpha) = \frac{cateto Adyacente}{hipotenusa}=\frac{b}{c}$$

* La tangente es la razón entre el cateto opuesto *lado a* y el cateto contiguo o cateto adyacente *lado b*.

$$tan( \alpha) = \frac{cateto Opuesto}{cateto Adyacente}=\frac{a}{b}$$


Se pueden observar estas funciones y su comportamiento en una gráfica como las siguientes:


```python
plt.figure(figsize=(10,6))
# Puntos a graficar
t = np.arange(0.1, 20, 0.1)

#Funciones
y1 = np.sin(t)
y2 = np.cos(t)

#Grafica
plt.plot(t, y1, label='Seno')
plt.plot(t, y2, label='Coseno')
plt.title('Seno y coseno')
plt.legend()

plt.show()
```

Observamos la función tangente en otra gráfica debido a los distintos rangos que tienen estas funciones, observemos que las funciones seno y coseno se encuentran en la imagen 0 y 1 para el eje y.

```python
# Puntos a graficar
t = np.arange(0.1, 20, 0.1)
plt.figure(figsize=(10,6))

#Funcion tangente
y3 = np.tan(t)

#Grafica
plt.plot(t, y3, label='Tangente')
plt.title('Tangente')
plt.show()
```

## Razones trigonométricas recíprocas

Las razones trigonométricas recíprocas de un ángulo α se obtienen como razones entre los tres lados de un triángulo rectángulo, siendo α uno de sus ángulos agudos.



*   Cosecante de α. Se define como la razón entre la hipotenusa *lado c* y el cateto opuesto *lado a*:

$$csc( \alpha) =\frac{1}{sen(\alpha)} \frac{hipotenusa}{cateto Opuesto}=\frac{c}{a}$$


*   Secante de α. Se define como la razón entre la hipotenusa *lado c* y el cateto adyacente *lado b*

$$sec( \alpha) =\frac{1}{cos(\alpha)} \frac{cateto Adyacente}{hipotenusa}=\frac{c}{b}$$

*    Cotangente de α. Se define como la razón entre el cateto contiguo o cateto adyacente  *lado b* y el cateto opuesto  *lado a*

$$cot( \alpha) =\frac{1}{tan(\alpha)} \frac{cateto Adyacente}{cateto Opuesto}=\frac{b}{a}$$

## Funciones trigonométricas inversas



*   El arcoseno es la función inversa del seno. Es decir:

Si $$arc sen(x) = sen^{-1}(x) = \alpha$$ entonces $$sen( \alpha) = x$$

*   El arcocoseno es la función inversa del coseno. Es decir:

Si $$arc cos(x) = cos^{-1}(x) = \alpha$$ entonces $$cos( \alpha) = x$$

*   El arcotangente es la función inversa de  la tangente. Es decir:

Si $$arc tan(x) = tan^{-1}(x) = \alpha$$ entonces $$tan( \alpha) = x$$

## Identidades trigonométricas
* Identidad fundamental de la trigonometría
$$sen^2 (\alpha) + cos^2 (\alpha) = 1 $$


* Relación entre el seno, coseno y tangente
$$tan (\alpha) = \frac{sen (\alpha)}{cos (\alpha)}$$

* Relación trigonométrica entre la tangente y la secante

$$ tan^2 (\alpha) + 1 = sec^2 (\alpha) $$

* Relación trigonométrica entre la cosecante y la cotangente
$$cot ^2(\alpha) + 1 = csc ^2 (\alpha)$$

## Razones trigonométricas de suma y resta de angulos

* Para la función seno tenemos que

$$ sen (\alpha \pm \beta) = sen(\alpha)cos(\beta) \pm cos(\alpha)sen(\beta)$$

* Para la función coseno tenemos que

$$ cos (\alpha \pm \beta) = cos(\alpha)cos(\beta) \mp sen(\alpha)sen(\beta)$$

* Para la función tangente tenemos que

$$ tan (\alpha + \beta) = \frac{tan(\alpha) +tan(\beta)}{1-tan(\alpha)tan(\beta)}$$

para el caso de la resta tenemos

$$ tan (\alpha - \beta) = \frac{tan(\alpha) -tan(\beta)}{1+tan(\alpha)tan(\beta)}$$

### Teorema de Pitagoras
Dado un triángulo rectángulo de catetos a y b e hipotenusa h (el lado opuesto al ángulo recto). Entonces:

$$h^2 = a^2 + b^2 $$

por lo que los lados estan denotados por:

$$ h = \sqrt{a^2 + b^2} $$

$$ a = \sqrt{h^2 - b^2} $$

$$ b = \sqrt{h^2 - a^2} $$


### Ejemplo 4

El ángulo óptimo de elevación $\theta$ de un lanzador de bala que tiene como objetivo el lanzamiento a la máxima distancia depende de la velocidad **v** del lanzamiento y de la altura inicial **h** desde la cual se lanza. Un modelo para que $\theta$ alcance la distancia máxima es

$$\theta = arcsen \left( \sqrt{\frac{v^2}{2v^2+64h}} \right)$$

image.png

Suponga que un lanzador de bala puede lanzar consistentemente una bola de acero donde h=6.6 pies y v =42 pies por segundo. ¿Cuál es el ángulo con el que debe lanzar la bala para obtener la distancia máxima?


Notemos que el modelo para obtener el ángulo puede ser calculado con facilidad, de modo que podemos obtener el ángulo necesario para el lanzador.

Realizamos los calculos con *Python*

```python
# parametros
v = 42
h = 6.6

#Aplicamos modelo
theta = math.sqrt((v**2) /((2*v**2) + 64*h))

#Usamos la funcion de grados
theta = math.degrees(np.arcsin(theta))
print('Los grados a los que debe ser lanzada son ', theta)
```

## Coordenadas polares

En un sistema de coordenadas polares, seleccionamos un punto llamado el polo y después, un rayo con un vértice en el polo, llamado eje polar.

```python
#Limites de la grafica
plt.figure(figsize=(10,6))
plt.xlim(-5, 5)
plt.ylim(-5,5)
#Ejes
plt.plot([-5,5],[0,0],color='black', linewidth=0.5)
plt.plot([0,0],[-5,5],color='black',linewidth=0.5)
#Polo
plt.scatter(0,0,marker="o",color='black',s=50)
#Eje polar
plt.arrow(0,0,4,0,width=0.1, ec='red')
plt.text(-1,0, 'polo',size=15)
plt.text(1.2,-0.6, 'eje polar',size=15)
plt.title('Sistema polar')
plt.show()
```

Un punto P en un sistema de coordenadas polares se representa por un par ordenado de números (r, $\theta$). Si r > 0, entonces r es la distancia del punto desde el polo; $\theta$ es un ángulo (en grados o en radianes) formado por el eje polar y un rayo desde el polo que pasa por el punto. Al par ordenado (r,$\theta$) le llamamos coordenadas polares del punto.

## Transformación de coordenadas polares a coordenadas rectangulares

A veces necesitamos convertir coordenadas o ecuaciones en forma rectangular a forma polar y viceversa. Para hacer esto, recuerda que el origen en las coordenadas rectangulares es el polo en las coordenadas polares y que el eje positivo x en las coordenadas rectangulares es el eje polar en las coordenadas polares.

Por lo que podemos denotar las coordenadas rectangulares de la siguiente manera:

$$ x = r cos \theta$$
$$ y = r sen \theta $$

De manera contraria encontramos que para convertir de coordenadas polares a coordenadas rectangulares tenemos:
$$r = \sqrt (x^2 + y^2)$$ 
y por otro lado 
$$\theta = tan^{-1} \left( \frac{y}{x} \right)$$
