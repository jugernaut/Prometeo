### Depreciación de electrodomésticos

Una licuadora nueva tiene un valor de $\$3,500$  y se deprecia linealmente $\$245$ anuales, se desea saber cuanto costará dentro de 5 años por lo que es necesario encontrar la recta que representa la depreciación anual de los electrodomésticos.

**Solución:** 
Dado que en la fecha actual la licuadora tiene un precio de  $\$3,500$ tendremos las coordenadas $(0,3500)$ y para el año siguiente costará $3,500 - 245=3,255$, por lo que podemos graficar la recta.


```python
#Definimos las coordenadas X y Y
x = [0,1] 
y = [3500,3255]

#Graficamos especificando el titulo y nombre de los ejes
par = [{'title':'Depreciación de una licuadora', 
        'xlabel':'$Año$',
        'ylabel':'$Precio$'}]

#Gráfica de la recta
graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=y)       
graf.show()
```

Después de la revisión de los siguientes subtemas será sencillo encontrar que la ecuación de la recta es:

$$ y = -245x + 3500 $$

Para el quinto año el valor de la licuadora será el siguiente:
$$  y = -245 (5) + 3500 = 2,275$$ 

## Pendiente de una recta
Se le da el nombre de pendiente $m$ a la tangente trigonométrica de la inclinación de la recta. De modo que,
$$ m = tan (\alpha) $$

donde $\alpha$ es el ángulo formado entre la recta y el eje de las abcisas.

Otra forma de ver y calcular la pendiente $m$ de una recta no vertical que pasa por dos puntos dados, está dada por el cociente de la diferencia de sus ordenadas y de la diferencia de sus abscisas, tomadas en el mismo orden.

$$m= \frac{\bigtriangleup y}{\bigtriangleup x} = \frac{y_2 - y_1}{x_2 -x_1}$$


### Ejemplo 1

Sacar la pendiente de la recta que pasa por los puntos $A(6,-1)$ y $B(-3,2)$ 

Primero graficaremos la recta que pasa por ambos puntos.

```python
#Definimos las coordenadas X y Y
x = [6,-3] 
y = [-1,2]

#Graficamos especificando el título y nombre de los ejes
par = [{'title':'Recta de los puntos $(6,-1)$ y $(-3,2)$', 
        'xlabel':'$x$',
        'ylabel':'$y$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=y)       #Gráfica de la recta
graf.scatter(x=6, y=-1,   #Punto A
             par={'label':'Punto A','color':'red', 'zorder':10})
graf.scatter(x=-3, y=2,   #Punto B
             par={'label':'Punto B','color':'yellow', 'zorder':10})
graf.legend()
graf.show()
```

Y calculamos la pendiente de la recta de la siguiente manera:

```python
#Definimos los puntos
x1 = 6
x2 = -3
y1 = -1
y2 = 2

#Calculamos la pendiente
m = (y2-y1)/(x2 - x1)

print('La pendiente de la recta es :', m)
```

### Ejemplo 2
Una imprenta en 2015 valía $\$25,500$, pero cuatro años antes, su valor era de $\$38,400$. Si el valor de la imprenta varía linealmente con el tiempo: 
  * a ) ¿Cuánto varía su valor por año? 
  * b ) ¿Cuál será su valor en 2017? 
  
**Solución**

En esta situación, la pendiente nos indica la variación del valor de la imprenta por año de uso, así que calculemos su valor, tomando como puntos $A(2011, 38400)$ y $B(2015, 25500)$.

```python
#Definimos las coordenadas X y Y
x = [2011,2015] 
y = [38400,25500]

#Graficamos especificando el titulo y nombre de los ejes
par = [{'title':'Recta sobre el valor de la imprenta', 
        'xlabel':'$Año$',
        'ylabel':'$Valor$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=y)       #Gráfica de la recta
graf.scatter(x=2011, y=38400,   #Punto A
             par={'label':'Punto A','color':'red', 'zorder':5})
graf.scatter(x=2015, y=25500,   #Punto B
             par={'label':'Punto B','color':'yellow', 'zorder':5})
graf.legend()
graf.show()
```

Y calculamos la pendiente de la recta de la siguiente manera:

```python
#Definimos los puntos
x1 = 2011
x2 = 2015
y1 = 38400
y2 = 25500

#Calculamos la pendiente
m = (y2-y1)/(x2 - x1)

print('La pendiente de la recta es :', m)
```

Por lo tanto; se debe interpretar que la imprenta pierde $\$ 3225$ cada año por uso y desgaste de sus partes.

Para conocer su valor en 2017, notemos que ya conocemos la devaluación anual por lo que se calcula de la siguiente manera:

$$  f(2017) = 25500 - 2(3225)$$

Pues estaremo aplicando la devaluación despues de 2 años. 
Por lo que el valor de la imprenta en el 2017 es:
$$ f(2017) = 19050$$

## Ecuación de la recta

En el ejercicio anterior encontramos la notacion de $f(x)$, pero ¿Cómo podemos encontrar aquella ecuación que represente la recta?

Es posible encontrar dicha ecuación según los requerimientos con los que se cuenten, por ejemplo, si tenemos un punto de la recta y su pendiente podemos obtener la ecuación como:
$$y -y_1 = m (x -x_1) $$

Cuando una recta corta algún eje en un punto, por ejemplo, el punto $A(a,0)$ se le conoce como abcisa al origen y al punto $B(0,b)$ se le conoce como ordenada al origen. Por lo que de la ecuación anterior, si tenemos la pendiente $m$ y una ordenada al origen obtenemos:

$$y = mx + b$$

Y es llamada ecuación pendiente-ordendada al origen.

La forma general de la ecuación de una recta es la siguiente:
$$ Ax +By +C = 0$$
De la cual mediante pasos algebráicos es posible obtener la forma pendiente ordenada-al origen. Y podemos deducir las siguientes propiedades:

* Si $A=0$, entonces $By + C=0$ representa una recta horiontal.
* Si $B=0$, entonces $Ax +C=0$ representa una recta vertical.
* Si $C=0$, entonces $Ax +By=0$ representa una recta que pasa por el origen del plano cartesiano.

## Intersección de rectas

Para encontrar la intersección entre dos rectas, es necesario resolver el sistema de ecuaciones dado por las ecuaciones de cada recta.



### Ejemplo 5
Se tienen las siguientes rectas:

Recta 1: $x+y -15= 0$

Recta 2: $7x+17y +65= 0$

**Solución:** 

Se resuelve el sistema de ecuaciones de forma que:

\begin{align}
7x+7y &= 105\\
7x+17y &=65\\
-10y&=170
\end{align}

Por lo tanto $y=\frac{170}{-10} = -17$

Ahora despejando la ecuación de la recta 1 tenemos:

\begin{align}
x+y -15&=0 \\
&= x -17 -15 = 0\\
&=x -32 =0\\
\end{align}

Por lo que $x=32$

Ya que tenemos ambos puntos las rectas se intersectan en el punto $(32,-17)$.

```python
#Definimos la función f(x) y g(x)
f = lambda x: -x + 15
g = lambda x: (-7*x - 65)/17

#Definimos el dominio y el codominio de la función f(x) y g(x)
x = np.linspace(-25,50,100)  # Dominio
yf = f(x)                 # Codominio de f
yg = g(x)                 # Codominio de g

#Graficamos especificando el título y nombre de los ejes
par = [{'title':'Intersección entre dos rectas', 
        'xlabel':'$x$',
        'ylabel':'$y$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=yf, par={'label':'$x+y -15= 0$'})                #Gráfica de la función f(x)
graf.plot(x=x, y=yg, par={'label':'$7x+17y +65= 0$'})                #Gráfica de la función g(x)


x0 = 32.0 # Modifique este valor entre 0 y 2

graf.scatter(x=x0, y=-17,         #Intersección
             par={'label':'Intersección', 's':50, 'marker':'s','color':'black', 'zorder':5})

graf.legend()
graf.show()
```

## Distancia de un punto a una recta.

Para calcular la distancia dirigida de un punto a una recta se tiene la siguiente ecuación

$$ d =  \frac{Ax_1+By_1 +C}{\pm \sqrt (A^2 + B^2)}$$

Seleccionando en el radical el mismo signo que $C$.

Hay que notar que la distancia resulta positiva cuando el punto se encuentra encima de la recta y negativa en caso contrario. 

Cuando se tiene la distancia no dirigida se toma el valor absoluto de la distancia dirigida.

### Ejemplo 6
Para el punto $P = (1, 2)$ y la recta $3x+4y-21=0$  determinar la distancia dirigida del punto a la recta.

**Solución:**

Utilizando la forma de la distancia anterior y sustituyengo los valores del ejercicio tenemos:

\begin{align}
d &=  \frac{Ax_1+By_1 +C}{\pm \sqrt (A^2 + B^2)}\\
&=  \frac{3x_1+4y_1 -21}{\pm \sqrt (3^2 + 4^2)}\\
&=  \frac{3(1)+4(2) -21}{\pm \sqrt (3^2 + 4^2)}\\
&= \frac{-10}{5}\\
&= -2
\end{align}

Por lo tanto la distancia dirigida al punto es $-2$.

Veamos la gráfica de este punto y la recta.

```python
#Definimos la función f(x) y g(x)
f = lambda x: (-3*x + 21)/4

#Definimos el dominio y el codominio de la función f(x) y g(x)
x = np.linspace(-2,5,100)  # Dominio
yf = f(x)                 # Codominio de f

#Graficamos especificando el titulo y nombre de los ejes
par = [{'title':'Distancia de un punto a una recta', 
        'xlabel':'$x$',
        'ylabel':'$y$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=yf, par={'label':'$3x+4y-21=0$'})       #Gráfica de la función f(x)

graf.scatter(x=1, y=2,         #Intersección
             par={'label':'Intersección', 's':50, 'marker':'s','color':'black', 'zorder':5})

#Distancia
x = [1,2] 
y = [2,3.7]
graf.plot(x=x, y=y, par={'label':'Distancia','ls':'--'})

graf.legend()
graf.show()
```

## Rectas en $\mathbb{R}^3$

Una recta en $\mathbb{R}^3$ puede verse como un plano paralelo al eje $z$. Para definir un plano es suficiente conocer un vector perpendicular al plano y un punto del mismo. Y para definir en forma vectorial una recta en   $\mathbb{R}^3$, es suficiente conocer un punto de la recta y un vector director que indique la dirección de la misma o sea un vector paralelo a la recta.


## Ecuación vectorial de la recta
Dados un vector $\vec{v}=(v_1,v_2,v_3)$ y un punto 
$P_0=(x_0,y_0,z_0)$, nos proponemos hallar la ecuación de la recta que pasa por el punto $P_0$ y es paralela al vector.

Consideremos un punto $P(x,y,z)$ perteneciente a la recta r. El vector $\vec{P_0P}$ resultará paralelo al vector director $\vec{v}$. Vemos que este vector será igual a 
$$\vec{P_0P} = \alpha \vec{v} $$
$$(x-x_0,y-y_0, z-z_0) = \alpha  (v_1,v_2,v_3)$$

Entonces se tendrá la ecuación

$$ (x,y,z) = (x_0,y_0,z_0) +\alpha  (v_1,v_2,v_3)$$

## Ecuación paramétrica de la recta

Derivado de la ecuación vectorial de la recta podemos encontrar la forma paramétrica de la recta de la siguiente manera:

$$x=x_0+\alpha v_1 $$
$$y=y_0+\alpha v_2 $$
$$z=z_0+\alpha v_3 $$

Con $\alpha \in \mathbb{R}$

## Ecuaciones simétricas de la recta

Dados $v_1,v_2,v_3$ distintos de cero las ecuaciones simétricas de la recta son:

$$\alpha = \frac{x-x_0}{v_1} = \frac{y-y_0}{v_2} =\frac{z-z_0}{v_3}$$



### Ejemplo 7 
A continuación graficamos un plano como ejemplo.

```python
from mpl_toolkits.mplot3d import axes3d

A = np.array([[-2, 5, 9],
              [ 7, 1, 1],
              [-3, 7,-1]])

B = np.array([1,6,-26])

ax = -5     # Intervalo X
bx = 5
ay = ax-2   # Intervalo Y
by = bx+2

# Ecuación de planos
z1 = lambda x,y: (-A[1,0]*x - A[1,1]*y + B[1])/A[1,2]

#Puntos X y Y
xi = np.linspace(ax,bx, 11)
yi = np.linspace(ay,by, 11)
Xi, Yi = np.meshgrid(xi,yi)
#Puntos en Z
Z1 = z1(Xi,Yi)

# GRAFICA de planos
figura = plt.figure(figsize=(10,8))
grafica = figura.add_subplot(111, projection='3d')

grafica.plot_wireframe(Xi,Yi,Z1, color ='green')

grafica.set_title('Ejemplo de un plano')
grafica.set_xlabel('x')
grafica.set_ylabel('y')
grafica.set_zlabel('z')
```
