Si se tienen dos espacios vectoriales $V$ y $W$ y tenemos una función o regla de asignación que transforma vectores de $V$ a vectores de $W$ no necesariamente será una transformación lineal, ya que debe cumplir la siguiente definición.


## Definición

Por definición una transformación lineal para los espacios vectoriales $V$ y $W$ sobre un campo $F$ es una función $T: V \rightarrow W $ tal que para todo $u,v \in V$:

*   $T(u+v) = T(u) +T(v)$.
*   $T(ku) = k T(u)$.





### Ejemplo 1

Verificar si la siguiente es una transformación entre los espacios $\mathbb{R}^2$ a  $\mathbb{R}$, por lo que tenemos $ T:  \mathbb{R}^2 \rightarrow \mathbb{R}$
$$ T(x,y)=x+y+1$$

**Solución:**

Verificando las condiciones para la definición de transformación:

Sean los puntos $T(1,1) = 3$ y $T(2,2)=5$

Por lo que para verificar la primer condición sobre la suma:
$$T(1+2,1+2)= T(3,3) = 7$$

Sin embargo;
$$T(1,1)+T(2,2) = 3+5= 8 $$
Y se pude ver que 
$$T(3,3) \neq T(1,1)+T(2,2)$$

Por lo que al no cumplir la primer condición $ T(x,y)=x+y+1$ no es una transformación lineal.

### Ejemplo 2 

También existen las transformaciones lineales al mismo espacio, veamos un ejemplo de ello.

Sea $T:\mathbb{R}^3 \rightarrow \mathbb{R}^3$
dada por
$$ T(x,y,z) = (2x,2y,2z) $$

**Solución:**

Para verificar que es una transformación lineal tomamos dos vectores $v_1,v_2 \in \mathbb{R}^3$ notando que $T(v)=2v$ tenemos

\begin{align}
T(v_1+v_2) &= 2(v_1+v_2)\\
&= 2(v_1)+2(v_2)\\
&= T(v_1)+T(v_2)
\end{align}

Así como 
\begin{align}
T(cv_1) &= 2(cv_1)\\
&= c(2v_1)\\
&= cT(v_1)
\end{align}

Por lo que $T$ es una transformación lineal.

## Propiedades de las transformaciones lineales

Las transformaciones lineales tienen las siguientes propiedades:

*   La imagen del vector nulo en $0_v$ es el vector nulo en el codominio $0_w$.
$$T(0_v) = 0_w$$
*   La imagen del vector $–v$ es igual al opuesto de la imagen de $v$.
$$T(-v)=-T(v) $$
*   Consideremos $r$ vectores del espacio vectorial $V$ como $v_1,v_2,...,v_r \in V$ tomando una combinación lineal
$$ \alpha_1 v_1 + \alpha_2 v_2+...+\alpha_r v_r$$

Sl aplicar la transformación lineal $T$,
$$ T(\alpha_1 v_1 + \alpha_2 v_2+...+\alpha_r v_r) = \alpha_1 T(v_1) + \alpha_2 T(v_2)+...+\alpha_r T(v_r)$$



Las transformaciones lineales pueden ir de un espacio vectorial a cualquier otro como veremos el en siguiente ejemplo:

### Ejemplo 3
Verificar si $T: M_{2,3}(\mathbb{R}) \rightarrow \mathbb{R}^3 $ la cual manda una matriz a un vector en $\mathbb{R}^3 $
\begin{equation}
M=
\begin{pmatrix}
a & b & c\\
d & e & f\\
\end{pmatrix}
\end{equation}

A un vector con la transformación

$$T(M)=(a-d,b-e,c-f)$$

**Solución:**

Para verificar que es una transformación lineal necesitamos determinar si se cumplen las primeras dos propiedades de la definición:

Para la primer propiedad tenemos que tomar dos matrices: 
\begin{equation}
M_1=
\begin{pmatrix}
a_1 & b_1& c_1\\
d_1 & e_1 & f_1\\
\end{pmatrix}
\end{equation}

\begin{equation}
M_2=
\begin{pmatrix}
a_2 & b_2 & c_2\\
d_2 & e_2 & f_2\\
\end{pmatrix}
\end{equation}
Realizando la suma de ambas matrices.

\begin{equation}
M_1 + M_2=
\begin{pmatrix}
a_1+ a_2 & b_1 +b_2 & c_1+c_2\\
d_1+d_2 & e_1+e_2 & f_1+f_2\\
\end{pmatrix}
\end{equation}
Así que aplicando la transformación tenemos:

$$T(M_1+M_2)=((a_1+ a_2)-(d_1+d_2),(b_1+b_2)-(e_1+e_2),(c_1+c_2)-(f_1+f_2))$$

Ahora por otro ado aplicando la transformación a las matrices por separado obtenemos:
$$T(M_1)=(a_1-d_1,b_1-e_1,c_1-f_1)$$
$$T(M_2)=(a_2-d_2,b_2-e_2,c_2-f_2)$$

Y realizamos la suma: 
$$T(M_1)+T(M_2)=((a_1+ a_2)-(d_1+d_2),(b_1+b_2)-(e_1+e_2),(c_1+c_2)-(f_1+f_2))$$

Por lo tanto obtenemos que 
$$T(M_1+M_2)=T(M_1)+T(M_2)$$

Para terminar vemos que se cumple la segunda condición

$$T(\alpha M)=(\alpha a-\alpha d,\alpha b-\alpha e,\alpha c-\alpha f)$$

Y por otro lado tenemos que

\begin{align}
\alpha T(M)&=alpha(a-d,b-e,c-f)\\
&=(\alpha a-\alpha d,\alpha b-\alpha e,\alpha c-\alpha f)
\end{align}

por lo tanto 
$$ T(\alpha M) = \alpha T(M) $$

Dado que hemos verificado la condiciones de la definición, se concluye que sí es una transformación lineal.

## Matriz asociada a una transformación lineal

Definiremos la matriz asociada a una transformación lineal $T: V \rightarrow W$ con $dim(V)=n$ y $dim(W)=m$ con bases en cada uno de los respectivos espacios:
$$B= {v_1,v_2,v_3,...,v_n}$$ 
$$B'= {w_1,w_2,w_3,...,w_m}$$ 
se define la matriz asociada $M(T)_{BB'}$ como la matriz respecto a las bases $B$ Y $B'$

$$M(T)_{BB'} = ([T(v_1)]_{B'} [T(v_2)]_{B'}... [T(v_n)]_{B'})$$

La cual se construye por columnas transformando los vectores de la base $B$ (del dominio) y expresando los transformados en sus coordenadas en la base $B'$. Por lo que tendrá $n$ columnas correspondientes a la base $B$ y $m$ filas correspondientes a la base $B'$, entonces se tendrá una matriz $$ M(T)_{BB'} \in \mathbb{R}^{mxn}$$


##Propiedades de la matriz asociada a una transformación lineal

La matriz asociada a una transformación lineal tiene la siguiente propiedad:

$$M(T)_{BB'}[v]_B=[T(v)]_{B'}$$

### Ejemplo 4

Sea la transformación lineal 
$$ T: \mathbb{R}^2 \rightarrow  \mathbb{R}^3| T((x,y))=(x+2y, x-y,y) $$

Consideremos las bases 
$$B=\{(0,1),(1,1)\}$$
$$B'=\{(1,0,0),(1,1,0),(0,1,1)\}$$

Para construir su matriz asociada a la transformación $T$ de la base $B$ a la base $B'$ tomamos:

$$T((0,1))=(2,-1,1)=\alpha (1,0,0)+\beta(1,1,0)+\gamma(0,1,1)$$
Por lo que aplicando la transformación,

\begin{equation}
[T((0,1))]_{B'} =
\begin{pmatrix}
4\\
-2\\
1
\end{pmatrix}
\end{equation}
Para el elemento $(1,1)$ de la base $B$
$$T((1,1))=(3,0,1)=\alpha (1,0,0)+\beta(1,1,0)+\gamma(0,1,1)$$
Por lo que aplicando la transformación,

\begin{equation}
[T((1,1))]_{B'} =
\begin{pmatrix}
4\\
-1\\
1
\end{pmatrix}
\end{equation}

Por lo tanto la matriz asociada queda:
\begin{equation}
M(T)_{BB'} =
\begin{pmatrix}
4 & 4\\
-2 & -1\\
1 & 1
\end{pmatrix}
\end{equation}




##Kernel de una transformación lineal

El kernel o núcleo de una transformación lineal $T: V \rightarrow W$ es el conjunto de vectores en $V$ que se van al vector $0$ de $W$ después de aplicar la transformación.

Es decir,

$$ ker(T) = \{v \in V: T(v)=0\}$$


### Ejemplo 5
Encontrar el kernel de la siguiente transformación $T: M_2(\mathbb{R}) \rightarrow M_{2,3}(\mathbb{R}) $

**Solución:**

\begin{equation}
T
\begin{pmatrix}
a & b\\
c & d\\
\end{pmatrix}
= 
\begin{pmatrix}
a & b\\
c & d\\
\end{pmatrix}
\begin{pmatrix}
1 & 1 & 1\\
1 & 1 & 1\\
\end{pmatrix}
\end{equation}

Por lo que buscamos aquel conjunto de elementos $A$ tal que $T(a)=0$  con $a\in A$ para esto dada la transformación es suficiente que los elementos de la matriz cumplan $a+b=0$ y $c+d=0$. Definiendo dicho kernel como:
\begin{equation}
ker(T)= \left\{
\begin{pmatrix}
a & -a\\
c & -c\\
\end{pmatrix}
:a,c \in \mathbb{R}
\right\}
\end{equation}


## Imagen de una transformación lineal
La imagen de $T: V \rightarrow W$ transformación lineal son los vectores en $W$ que se puede escribir de la forma $T(v)$ para algún $v \in V$, es decir, es la imagen en el sentido clásico de teoría de conjuntos o de cálculo.

$$Im(T) = \{T(v): v \in V\}$$

##Rango de una transformación lineal

Se define que el rango es igual a la dimensión de la imagen de la transformación lineal. Es decir,
$$rango(M(T)_{BB'}=dim(Im(T)) $$
