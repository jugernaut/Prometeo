### Introducción

**¿Sabías que los juegos para niños, pueden verse como vectores?**\
Cuando se tienen 2 o más columpios, estos pueden verse como vectores paralelos, ya que estos se mueven en una misma dirección, más no en un mismo sentido, ya que el movimiento de ellos depende de la fuerza que se les aplique y el tiempo en el que se aplica dicha fuerza. Dentro de este notebook, aprenderemos ¿Qué es un vector?, así algunas operaciones con estos.

El término vector proviene del latín vector, vectoris, cuyo significado es 'el que conduce', o 'el que transporta'.\
Se le denomina vector de dimensión $n$ a una tupla de números reales (Componentes del vector), El conjunto de todos los vectores de dimensión, se representa como $R^n$. Es decir, un vector $v$ perteneciente a un espacio $R^n$ se representa de la siguiente manera:\
$v=(a_{1},a_{2},a_{3},...,a_{n})$, donde $v \in R^n$

   + **¿Qué es un vector?**\
\
Dentro del área de la física, se le conoce como vector a un segmento de recta en el espacio que parte de un punto hacia otro, es decir, que tiene dirección y sentido. Los vectores en física tienen por función expresar las llamadas magnitudes vectoriales.\
En matemáticas se define vector como un elemento de un espacio vectorial. Esta noción es más abstracta y para muchos espacios vectoriales no es posible representar sus vectores mediante el módulo y la dirección. En particular los espacios de dimensión infinita sin producto escalar no son representables de ese modo. Los vectores en un espacio euclídeo se pueden representar geométricamente como segmentos de recta $R$, en el plano $R^{2}$ (bidimensional), o en el espacio $R^{3}$ (tridimensional).\
Mientras en el área de la computación, los vectores son una forma de almacenar datos que permiten contener una serie de valores del mismo tipo, cada uno de los valores contenidos tiene una posición asociada que se usará para accederlos. Está posición o índice será siempre un número entero positivo.



Mediante la librería $Numpy$, podemos generar arreglos, los cuales vistos de otro modo pueden ser vectores.


   + **Suma y resta de vectores**\
\
Para sumar dos vectores $a$  y $b$ se suma $a$ con el vector $b$, es decir, se suman los componentes de cada vector:
\begin{equation}
a+b = (a_{x}+b_{x},a_{y}+b_{y},a_{z}+b_{z})
\end{equation}
Por ejemplo:\
Sean $a=(3,2,-4)$ y $b=(-3,2,7)$, calcula el vector $a+b$
\begin{equation}
a+b = (3+(-3),2+2,-4+7)=(0,4,3)
\end{equation}
\
\
Para restar dos vectores $a$ y $b$ se suma $a$ con el opuesto del vector $b$, es decir:}
\begin{equation}
a-b = (a_{x}-b_{x},a_{y}-b_{y},a_{z}-b_{z})
\end{equation}
Por ejemplo:\
Sea $a=(5,2,4)$ y $b=(-3,5,9)$, calcular el vector $a-b$
\begin{equation}
a-b=(5-(-3),2-5,4-9)=(8,-3,-5)
\end{equation}

Sumando y Restando Vectores, al vector $x$ se le desea sumar otro vector $y$

 + **Multiplicación de un Vector por un escalar**\
\
Al multiplicar un vector $x$ por un escalar $\lambda$, obtenemos como un nuevo vector $y=(\lambda)(x)$ que cumple con las siguientes características:\
Si $\lambda$ es:\
\
~ Positivo $x$ y $y$ tendrán el mismo sentido.\
~ Negativo $x$ y $y$ tendrán diferente sentido.\
\
Por ejemplo:\
Sea $x =(2,4,1,2)$ y $\lambda=2$\
\begin{equation}
(x)(\lambda) = 2(2,4,1,2) = (4,8,2,4)
\end{equation}

Para multiplicar un vector $x$ por un escalar $\lambda$, únicamente definimos dicho escalar y realizamos el producto con nuestro arreglo.

```
k = 2
vector = np.array([2, 4, 1, 2])
producto = vector*k
print('Vector: ', vector, '\n')
print('Escalar: ', k, '\n')
print(f'El producto es: {producto}')
```

Cuando el escalar es positivo

```
par = [{'title':'Producto (+Escalar)(Vector)','ylim':(-3,9), 'xlim':(-0.5,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1, 2, 3]), y=vector, par={'label':'Vector', 'ls':'dashdot','lw':2, 'color':'blue'})
graf.plot(x=np.array([0, 1, 2, 3]), y= producto, par={'label':'Producto', 'ls':':','lw':3, 'color':'red'})
graf.scatter(x=np.array([0, 1, 2, 3]), y=vector, par={ 'ls':'dashdot','lw':3, 'color':'blue'})
graf.scatter(x=np.array([0, 1, 2, 3]), y=producto, par={ 'ls':':','lw':5, 'color':'red'})
graf.legend()
graf.show()
```

Cuando el escalar es negativo, se tienen direcciones distintas

```
k = -2
producto = vector*k
print('Escalar: ',k)
print(f'El producto es: {producto}')
```

```
par = [{'title':'Producto (-Escalar)(Vector)','ylim':(-10,6), 'xlim':(-0.5,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1, 2, 3]), y=vector, par={'label':'Vector', 'ls':'dashdot','lw':2, 'color':'blue'})
graf.plot(x=np.array([0, 1, 2, 3]), y= producto, par={'label':'Producto', 'ls':':','lw':3, 'color':'red'})
graf.scatter(x=np.array([0, 1, 2, 3]), y=vector, par={ 'ls':'dashdot','lw':3, 'color':'blue'})
graf.scatter(x=np.array([0, 1, 2, 3]), y=producto, par={ 'ls':':','lw':5, 'color':'red'})
graf.legend()
graf.show()
```

### Vector Unitario
Un vector unitario hace referencia al vector cuyo módulo es igual a 1. Cabe recordar que el módulo es la cifra coincidente con la longitud cuando el vector se representa en un gráfico. El módulo, de este modo, es una norma de la matemática que se aplica al vector que aparece en un espacio euclídeo.
Para hallar un vector unitario a partir de cualquier vector, hay que dividir este último por su módulo.
### Ejercicio:
Si $v$ es un vector de componentes (3,4), encontrar un vector unitario de su misma dirección y sentido, dada la fórmula:
\begin{equation}
u = \frac{1}{|v|} v
\end{equation}
- Calcular magnitud de $v$
\begin{equation}
|v| = \sqrt{(3)^2 + (4)^2} = \sqrt{25} = 5
\end{equation}
Sustituyendo:
\begin{equation}
u = \frac{1}{5}(3,4) = (\frac{3}{5},\frac{4}{5})
\end{equation}
Visto desde código:

```
vector_unitario = []
vector = [3,4]
#Calculando magnitud
magnitud = math.sqrt(vector[0]**2 + vector[1]**2)
for i in range(len(vector)):
  vector_unitario.append(vector[i]*(1/magnitud))
print('Nuestro vector unitario es: ',vector_unitario)
```

```
par = [{'ylim':(0,5), 'xlim':(-0.25,1.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1]), y=vector, par={'label':'Vector', 'ls':'dashdot','lw':2, 'color':'blue'})
graf.plot(x=np.array([0, 1]), y= vector_unitario, par={'label':'Vector Unitario', 'ls':':','lw':3, 'color':'red'})
graf.scatter(x=np.array([0, 1]), y=vector, par={ 'ls':'dashdot','lw':3, 'color':'blue'})
graf.scatter(x=np.array([0, 1]), y=vector_unitario, par={ 'ls':':','lw':5, 'color':'red'})
graf.legend()
graf.show()
```

# Ejercicio
En un juego de futbol americano, un mariscal de campo toma el balón desde la línea de golpeo, corre hacia atrás una distancia de $10$ yardas y luego corre de manera lateral, paralelo a la línea de golpeo $15$ yardas. En este punto lanza un pase recto hacia adelante $50$ yardas, perpendicular a la línea de golpeo, ¿Cual es la magnitud de desplazamiento?

```
par = [{'ylim':(41,-11), 'xlim':(-0.25,16)}]
vectores = [[15, 40], [15, -10]]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 15]), y=np.array([0, 40]), par={'label':'', 'ls':'dashdot','lw':2, 'color':'blue'})
graf.plot(x=np.array([0, 15]), y= np.array([0, -10]), par={'label':'', 'ls':':','lw':3, 'color':'red'})
graf.scatter(x=np.array([0, 15]), y=np.array([0, 40]), par={ 'ls':'dashdot','lw':3, 'color':'blue'})
graf.scatter(x=np.array([0, 15]), y= np.array([0, -10]), par={ 'ls':':','lw':5, 'color':'red'})
graf.show()

```

```
dy = -10 + 50
dx = 15
dt = dy + dx
d = math.sqrt((15**2)+(40**2))
print('La magnitud del desplazamiento resultante del balón es: ', d, ' yardas')
```

Comprobando

\begin{equation}
\sum dy = -10+50 = 40
\end{equation}
\
\begin{equation}
\sum dx = 15
\end{equation}
\
\begin{equation}
dt = 15yardas + 40 yardas
\end{equation}
\
$$|d| = \sqrt{(15)^2 + (40)^2 }$$
\
$$|d| = \sqrt{225 + 1600}$$
\
$$|d| = \sqrt{1825}$$
\
$$|d| = 42.72 yardas$$
\
$\therefore$ La magnitud del desplazamiento del balón es de $42.72$ yardas
