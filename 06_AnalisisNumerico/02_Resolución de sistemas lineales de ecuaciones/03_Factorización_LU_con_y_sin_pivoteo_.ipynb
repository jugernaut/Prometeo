###Eliminación Gaussiana y factorización LU

Recordemos que el producto $L_kL_j$ resulta una matriz triangular inferior
con diagonal unitaria si k < j por lo tanto:

\begin{equation} L=M^{-1}=(M_{n-1} M_{n-2}...M_1)^{-1} = M_1^{-1}...M_{n-1}^{-1}=L_1...L_{n-1}  \end{equation}

L es una matriz triangular inferior con diagonal unitaria. Hemos visto
que por diseño la matriz U = MA es triangular superior. Por lo tanto
la matriz A se puede expresar como un producto de una matriz
triangular inferior por una triangular superior:

\begin{equation} A=LU  \end{equation}

Esta factorización del sistema de ecuaciones original Ax = b ayuda a
escribirlo como:

\begin{equation} LUx=b  \end{equation}

Una vez que tenemos el sistema de ecuaciones en su forma LU para
resolverlo:

- Resolvemos primero el sistema triangular inferior \begin{equation} Ly=b  \end{equation} mediante sustitución hacia adelante

- Una vez que conocemos el vector y resolvemos el sistema triangular superior \begin{equation} Ux=y  \end{equation} mediante sustitución hacia atrás. Nótese que la solución intermedia $y$ es la misma que el vector transformado Mb.

La eliminación Gaussiana y la factorización LU son dos formas de
expresar el mismo proceso de solución.

Observe que en el proceso descrito anteriormente, la matriz $A^{(n-1)}$ es una matriz triangular superior, además cada matriz $M_k$ es triangular inferior unitaria, por lo que podemos definir la matrix L de la siguiente forma:

\begin{equation} L^{-1} = \prod_{k=1}^{n-1}M_{n-k}=M_{n-1}M_{n-2}...M_2M_1
\end{equation}

la cual es una matriz (Triangular Inferiror Unitaria=TIU) y por ende “no singular”, lo que implica que es invertible y su inversa será TIU.

Observe también que si denotamos $A^{(n-1)}$ = U, podemos escribir:

\begin{equation} U =  L^{-1}A \end{equation}
\begin{equation} \Rightarrow (L^{-1})^{-1}U = (L^{-1})^{-1} L^{-1}A \Rightarrow (L^{-1})^{-1} U = A
\end{equation}

y tomando $L = (L^{-1})^{-1}$, estaremos generando la factorización

\begin{equation} A =  LU \end{equation}

donde

\begin{equation} U = A^{n-1} \hspace{.5 cm} y \hspace{.5 cm} L = (M_{n-1} M_{n-2} ...M_2M_1)^{-1} = M_1^{-1}M_2^{-1}...M_{n-2}^{-1}M_{n-1}^{-1} \end{equation}

Dada una matriz triangular inferior elemental $M_i$ (Matriz de eliminación Gaussiana), su inversa se puede escribir como:

\begin{equation} M_i^{-1}=I-m_ie_i^T
\end{equation}

Donde $e_i= (0,...,0,1,0,...,0)$ y $m_1^T= (0,...0,m_{i+1,i},...,m_{n,i})$. Por lo cual, es fácil seguir que:

###$\color{blue}{Teorema (Existencia\hspace{.1 cm} y\hspace{.1 cm} unicidad\hspace{.1 cm} de\hspace{.1 cm} la\hspace{.1 cm} factorización\hspace{.1 cm} LU)}$

(i) Una matriz A de tamaño $n \times n, n > 1$ tiene factorización LU si $A^k$ es no singular, $\forall k = 1,..., n-1$.

(ii) Si A tiene factorización LU y además A es no singular, entonces la factorización es única.

Prueba (I): Inducción sobre $n$.

(Para $n=2$), sea

\begin{equation} A=
\begin{bmatrix} 
a & b\\
c & d
\end{bmatrix} , a \not = 0
\end{equation}

Observe que 
\begin{equation} A=
\begin{bmatrix} 
1 & 0\\
\mu & 1
\end{bmatrix} \begin{bmatrix} 
\alpha & \beta\\
0 & \gamma
\end{bmatrix}
\end{equation}

donde $\mu = \frac{c}{a},\hspace{.4 cm} \alpha = a,\hspace{.4 cm} \beta=b,\hspace{.4 cm} \gamma = d-\frac{bc}{a} \hspace{.4 cm}$ todas bien definidas

(HI) $A \in M_n (\mathbb R)$, con $A^{(k)}$ no singulares para $1 \leq k \leq n$, entonces A tiene factorización LU.

(Caso $n+1$) Sea $A \in M_{n+1} (\mathbb R)$, con $A^{(k)}$ no singular para $1 \leq k \leq n+1$ debemos probar que A tiene también factorización LU. En efecto, primero consideremos la matriz A de la forma:

<center>
<img src="https://github.com/jugernaut/Prometeo/blob/desarrollo/Figuras/EJEA_LU_8.PNG?raw=1" width="200"> 
</center>

observe que por hipótesis $A^{(n)}$ tiene todos sus menores principales no singulares, entonces por $HI$ existen $U^{(n)}$ triangular superior y $L^{(n)}$
triangular inferior unitaria tal que, $A^{(n)} = L^{(n)}U^{(n)}$, entonces para escribir $A = LU$ basta tomar

Observe que 
\begin{equation} L=
\begin{bmatrix} 
L^{(n)} & 0\\
\mu^T & 1
\end{bmatrix} \hspace{.8 cm} y \hspace{.8 cm} U= \begin{bmatrix} 
U^{(n)} & \beta\\
0 & \gamma
\end{bmatrix}
\end{equation}

donde

$\mu^T = c^T(U^{(n)})^{-1},\hspace{.8 cm} \beta= (L^{(n)})^{(-1)}b,\hspace{.8 cm} \gamma= d-\mu^T \beta$

Prueba(II) Sea $A$ no singular, supongamos que $A = L_1U_1 = L_2U_2$, observe que $L_1, U_1, L_2, U_2$ son no singulares, entonces:

\begin{equation} L_1U_1=L_2U_2 \end{equation}
\begin{equation} (L_2)^{-1}L_1U_1=(L_2)^{-1}L_2U_2  \end{equation}
\begin{equation} (L_2)^{-1}L_1U_1(U_1)^{-1}=U_2(U_1)^{-1} \Rightarrow (L_2)^{-1}L_1 = U_2(U_1)^{-1}  \end{equation}

observe que $(L_2)^{-1}$ es $(TIU)$, y además $U_2(U_1)^{-1}$ es $(TS)$, por lo que la única opción es que ambos productos sean la identidad.

Entonces $L_1 = L_2$ y $U_1 = U_2$.

###Ejemplo de pivoteo y singularidades

Consideremos la matriz

\begin{equation} A=
\begin{bmatrix} 
0 & 1\\
1 & 0
\end{bmatrix}
\end{equation}

que es no singular, pero no tiene factorización $LU$ a menos que intercambiemos los renglones. Mientras que la matriz singular:

\begin{equation} A=
\begin{bmatrix} 
1 & 1\\
1 & 1
\end{bmatrix} = \begin{bmatrix} 
1 & 0\\
1 & 1
\end{bmatrix} = \begin{bmatrix} 
1 & 1\\
0 & 0
\end{bmatrix} = LU
\end{equation}  

la matriz triangular superior $U$ es singular y a pesar de que se podría
completar la factorización $LU$, fallará la sustitución hacia atrás (pues
la matriz $A$ es singular por lo tanto NO tiene factorizaciíon $LU$). En
principio cualquier valor diferente de cero puede ser un pivote para
calcular los multiplicadores, pero por la precisión aritmética finita la
elección debe hacerse con cuidado para minimizar el error.

###Pivoteo parcial

Los multiplicadores no deben de exceder de 1 en magnitud si en cada
columna elegimos la entrada en magnitud más grande en/o debajo de
la diagonal principal como pivote. Esto se conoce como pivoteo
parcial. Usando precisión finita debemos de evitar pivotes cero, pero
también pivotes muy pequeños. Consideremos

\begin{equation} A=
\begin{bmatrix} 
\epsilon & 1\\
1 & 1
\end{bmatrix} 
\end{equation}  

con $\epsilon$ un número positivo, pero muy pequeño, si no intercambiamos los
renglones entonces el pivote es $\epsilon$ y el multiplicador de la matriz de
eliminación es $-1/\epsilon$ quedando esta matriz como:

\begin{equation} M=
\begin{bmatrix} 
1 & 0\\
-1/\epsilon & 1
\end{bmatrix} 
\end{equation}  

de donde

\begin{equation} L=
\begin{bmatrix} 
1 & 0\\
1/\epsilon & 1
\end{bmatrix} y \hspace{.4 cm} U= \begin{bmatrix} 
\epsilon & 1\\
0 & 1-1/\epsilon
\end{bmatrix} = \begin{bmatrix} 
\epsilon & 1\\
0 & -1/\epsilon
\end{bmatrix} 
\end{equation}  

si $\epsilon < \epsilon_{mach} \Rightarrow 1/\epsilon$ es muy grande, por lo que $1-1/\epsilon$ esta dominado por $-1/\epsilon$. Pero entonces

\begin{equation} LU=
\begin{bmatrix} 
1 & 0\\
1/\epsilon & 1
\end{bmatrix} \begin{bmatrix} 
\epsilon & 1\\
0 & -1/\epsilon
\end{bmatrix} = \begin{bmatrix} 
\epsilon & 1\\
1 & 0
\end{bmatrix} \neq A
\end{equation}  

Usando un pivote pequeño y un multiplicador correspondiente grande
ha causado una pérdida de información que no puede recuperarse. Si
en cambio intercambiamos los renglones de $A$ tenemos que la matriz
de eliminación es

\begin{equation} M=
\begin{bmatrix} 
1 & 0\\
-\epsilon & 1
\end{bmatrix} 
\end{equation} 

\begin{equation} L=
\begin{bmatrix} 
1 & 0\\
\epsilon & 1
\end{bmatrix} y \hspace{.4 cm} U= \begin{bmatrix} 
1 & 1\\
0 & 1-\epsilon
\end{bmatrix} = \begin{bmatrix} 
1 & 1\\
0 & 1
\end{bmatrix} 
\end{equation}  

Tenemos que 

\begin{equation} LU=
\begin{bmatrix} 
1 & 0\\
\epsilon & 1
\end{bmatrix} \begin{bmatrix} 
1 & 1\\
0 & 1
\end{bmatrix} = \begin{bmatrix} 
1 & 1\\
\epsilon & 1
\end{bmatrix} 
\end{equation}  

que es un resultado correcto después de una permutación (intercambio
de renglones).

En general los pivotes más grandes producirán multiplicadores más
pequeños y por lo tanto, errores más pequeños. En particular la entrada
más grande en o por debajo de la diagonal es usada como pivote. Cada
matriz de eliminación $M_k$ es precedida por una matriz de permutación
$P_k$ que intercambia renglones para tener las entradas en magnitud más
grandes en o debajo de la diagonal en la columna $k$ en la posición de
pivote. Seguiremos teniendo que

\begin{equation} MA=U \end{equation}

donde $U$ es triangular superior, pero ahora

\begin{equation} M=M_{n-1}P_{n-1}...M_1P_1 \end{equation}

$M$ es triangular, pero en el sentido general. En este caso la
descomposición $LU$ no significa tri. inferior X tri. superior como antes, pero es igualmente útil.

###Factorización LU mediante eliminación Gaussiana y pivoteo parcial

Notemos que la matriz de permutación
\begin{equation} P=P_{n-1}...P_1 \end{equation}

permuta los renglones de $A$ en el orden determinado por el pivoteo
parcial. Entonces obtenemos la factorización
\begin{equation} PA=LU \end{equation}

donde $L$ es triangular inferior. Para resolver $Ax = b$, primero
resolvemos el sistema triangular inferior $Ly = Pb$ mediante
sustitución hacia adelante y después resolvemos el sistema triangular
superior $Ux = y$ mediante sustitución hacia atrás.

\begin{equation} Ax=
\begin{bmatrix} 
1 & 2 & 4\\
4 & 4 & 2\\
4 & 6 & 4
\end{bmatrix} \begin{bmatrix} 
x_1 \\
x_2 \\
x_3
\end{bmatrix} = \begin{bmatrix} 
3 \\
6 \\
10
\end{bmatrix} = b
\end{equation}  

La entrada más grande en la primera columna es $4$, intercambiamos los
primeros dos renglones usando la matriz de permutación $P$

\begin{equation} P_1=
\begin{bmatrix} 
0 & 1 & 0\\
1 & 0 & 0\\
0 & 0 & 1
\end{bmatrix}
\end{equation}  

obteniendo el sistema permutado

\begin{equation} P_1Ax=
\begin{bmatrix} 
4 & 4 & 2\\
1 & 2 & 2\\
4 & 6 & 4
\end{bmatrix} \begin{bmatrix} 
x_1 \\
x_2 \\
x_3
\end{bmatrix} = \begin{bmatrix} 
6 \\
3 \\
10
\end{bmatrix} = P_1b
\end{equation}

###Pivoteo Completo

El pivoteo parcial se llama así pues solamente se busca entre los
elementos de una columna para buscar el pivote de esa columna. La
técnica de pivoteo completo consiste en buscar la entrada más grande
en una submatriz, de donde se hace la permutación conveniente para
usar esa entrada como pivote. Esto requiere de intercambio tanto de
renglones como de columnas. Esto deja una factorización de la forma:

\begin{equation} PAQ=LU \end{equation}

donde $L$ es triangular inferior, $U$ es triangular superior y $P$ y $Q$ son
matrices de permutación que reordenan los renglones y las columnas
de $A$ respectivamente. Ahora para resolver el sistema de ecuaciones

\begin{equation} Ax=b \end{equation}

Primero resolvemos el sistema triangular inferior

\begin{equation} Ly=Pb \end{equation}

mediante sustición hacia adelante y después resolvemos el sistema

\begin{equation} Uz=y \end{equation}

mediante sustitución hacia atrás y finalmente permutamos los
componentes de la solución $x = Qz$. Aunque la estabilidad numérica
del pivoteo completo es teóricamente mejor, requiere de mucho más
esfuerzo computacional.
